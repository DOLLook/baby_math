```python
import matplotlib.pyplot as plt
import numpy as np

def draw_circle():
    """ÁªòÂà∂ÂúÜÂΩ¢"""
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # ÁªòÂà∂ÂúÜÂΩ¢
    circle = plt.Circle((0, 0), 1, facecolor='lightblue', 
                       edgecolor='blue', linewidth=4)
    ax.add_patch(circle)
    
    # Ê†áÊ≥®
    ax.text(0, 0, 'ÂúÜÂΩ¢', fontsize=20, ha='center', va='center', fontweight='bold')
    ax.text(0, -1.3, 'ÂúÜÂúÜÁöÑÔºåÊ≤°ÊúâËßí', fontsize=14, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# ÊòæÁ§∫ÂúÜÂΩ¢
print("üîµ ËÆ§ËØÜÂúÜÂΩ¢Ôºö")
print("ÂúÜÂΩ¢ÊòØÂúÜÂúÜÁöÑÔºåÊ≤°ÊúâËßí")
print("ÂÉèÂ§™Èò≥„ÄÅÊúà‰∫Æ„ÄÅËΩ¶ËΩÆÈÉΩÊòØÂúÜÂΩ¢")
draw_circle()
```

```python
def draw_square():
    """ÁªòÂà∂Ê≠£ÊñπÂΩ¢"""
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # ÁªòÂà∂Ê≠£ÊñπÂΩ¢
    square = plt.Rectangle((-0.5, -0.5), 1, 1, facecolor='lightgreen', 
                          edgecolor='green', linewidth=4)
    ax.add_patch(square)
    
    # Ê†áÊ≥®Ëßí
    corners = [(-0.5, 0.5), (0.5, 0.5), (0.5, -0.5), (-0.5, -0.5)]
    for i, (x, y) in enumerate(corners):
        ax.text(x, y + 0.15, f'Ëßí{i+1}', fontsize=12, ha='center', fontweight='bold')
    
    # Ê†áÊ≥®Ëæπ
    ax.text(0, 0.6, 'Ëæπ1', fontsize=12, ha='center', fontweight='bold')
    ax.text(0.6, 0, 'Ëæπ2', fontsize=12, ha='center', va='center', fontweight='bold')
    ax.text(0, -0.6, 'Ëæπ3', fontsize=12, ha='center', fontweight='bold')
    ax.text(-0.6, 0, 'Ëæπ4', fontsize=12, ha='center', va='center', fontweight='bold')
    
    # Ê†áÊ≥®
    ax.text(0, 0, 'Ê≠£ÊñπÂΩ¢', fontsize=20, ha='center', va='center', fontweight='bold')
    ax.text(0, -1.3, '4Êù°ËæπÔºå4‰∏™ËßíÔºåËæπËæπÁõ∏Á≠â', fontsize=14, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# ÊòæÁ§∫Ê≠£ÊñπÂΩ¢
print("üü© ËÆ§ËØÜÊ≠£ÊñπÂΩ¢Ôºö")
print("Ê≠£ÊñπÂΩ¢Êúâ4Êù°ËæπÔºå4‰∏™Ëßí")
print("ÊØèÊù°ËæπÈÉΩ‰∏ÄÊ†∑Èïø")
print("ÂÉèÊñπÂΩ¢ÁöÑÊ°åÂ≠ê„ÄÅÁ™óÊà∑ÈÉΩÊòØÊ≠£ÊñπÂΩ¢")
draw_square()
```

```python
import random

def find_shape_game():
    """ÊâæÂõæÂΩ¢Ê∏∏Êàè"""
    print("üéÆ ÊâæÂõæÂΩ¢Ê∏∏Êàè")
    print("=" * 40)
    print("ËßÑÂàôÔºöÁúãÂõæÂΩ¢ÔºåËØ¥Âá∫ÂÆÉÁöÑÂêçÂ≠ó")
    
    score = 0
    total_questions = 5
    
    shapes = [
        {'name': 'ÂúÜÂΩ¢', 'draw': lambda ax: ax.add_patch(plt.Circle((0, 0), 0.8, facecolor='lightblue', edgecolor='blue', linewidth=4))},
        {'name': 'Ê≠£ÊñπÂΩ¢', 'draw': lambda ax: ax.add_patch(plt.Rectangle((-0.4, -0.4), 0.8, 0.8, facecolor='lightgreen', edgecolor='green', linewidth=4))},
        {'name': '‰∏âËßíÂΩ¢', 'draw': lambda ax: ax.add_patch(plt.Polygon([(-0.4, -0.4), (0.4, -0.4), (0, 0.4)], facecolor='lightyellow', edgecolor='orange', linewidth=4))},
        {'name': 'ÈïøÊñπÂΩ¢', 'draw': lambda ax: ax.add_patch(plt.Rectangle((-0.6, -0.4), 1.2, 0.8, facecolor='lightpink', edgecolor='red', linewidth=4)))}
    ]
    
    for i in range(total_questions):
        print(f"\nÈóÆÈ¢ò {i+1}/{total_questions}")
        
        # ÈöèÊú∫ÈÄâÊã©‰∏Ä‰∏™ÂõæÂΩ¢
        shape = random.choice(shapes)
        
        # ÁªòÂà∂ÂõæÂΩ¢
        fig, ax = plt.subplots(figsize=(6, 6))
        shape['draw'](ax)
        ax.set_xlim(-1.2, 1.2)
        ax.set_ylim(-1.2, 1.2)
        ax.set_aspect('equal')
        ax.axis('off')
        plt.tight_layout()
        plt.show()
        
        # ËØ¢ÈóÆÁ≠îÊ°à
        answer = input("ËøôÊòØ‰ªÄ‰πàÂõæÂΩ¢Ôºü")
        
        if answer == shape['name']:
            print("‚úÖ Á≠îÂØπ‰∫ÜÔºÅ")
            score += 1
        else:
            print(f"‚ùå ‰∏çÂØπÂì¶ÔºåÊ≠£Á°ÆÁ≠îÊ°àÊòØ {shape['name']}")
    
    print(f"\nüéØ Ê∏∏ÊàèÁªìÊùüÔºÅ")
    print(f"‰Ω†ÁöÑÂæóÂàÜÔºö{score}/{total_questions}")
    
    if score == total_questions:
        print("üèÜ Â§™ÂéâÂÆ≥‰∫ÜÔºÅÂÖ®ÈÉ®Ê≠£Á°ÆÔºÅ")
    elif score >= total_questions * 0.8:
        print("üåü ÂÅöÂæóÂæàÂ•ΩÔºÅÁªßÁª≠‰øùÊåÅÔºÅ")
    else:
        print("üí™ ÁªßÁª≠Âä†Ê≤πÔºåÂ§öÁªÉ‰π†Â∞±‰ºöËøõÊ≠•ÔºÅ")

# ËøêË°åÊ∏∏Êàè
find_shape_game()
```

## 8. ÂõæÂΩ¢ËÆ°Êï∞

Êï∞‰∏ÄÊï∞ÊúâÂ§öÂ∞ë‰∏™ÂõæÂΩ¢ÔºÅ

```python
def count_shapes():
    """ÂõæÂΩ¢ËÆ°Êï∞"""
    print("üî¢ ÂõæÂΩ¢ËÆ°Êï∞")
    print("=" * 40)
    
    # ÂàõÂª∫ÈöèÊú∫ÂõæÂΩ¢
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # ÁîüÊàêÈöèÊú∫ÂõæÂΩ¢
    shapes = []
    for _ in range(10):
        shape_type = random.choice(['circle', 'square', 'triangle', 'rectangle'])
        x = random.uniform(-3, 3)
        y = random.uniform(-3, 3)
        
        if shape_type == 'circle':
            circle = plt.Circle((x, y), 0.3, facecolor='lightblue', 
                               edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
            shapes.append('ÂúÜÂΩ¢')
        elif shape_type == 'square':
            square = plt.Rectangle((x-0.25, y-0.25), 0.5, 0.5, facecolor='lightgreen', 
                                  edgecolor='green', linewidth=2)
            ax.add_patch(square)
            shapes.append('Ê≠£ÊñπÂΩ¢')
        elif shape_type == 'triangle':
            triangle = plt.Polygon([(x-0.3, y-0.25), (x+0.3, y-0.25), (x, y+0.25)], 
                                  facecolor='lightyellow', edgecolor='orange', linewidth=2)
            ax.add_patch(triangle)
            shapes.append('‰∏âËßíÂΩ¢')
        elif shape_type == 'rectangle':
            rectangle = plt.Rectangle((x-0.4, y-0.25), 0.8, 0.5, facecolor='lightpink', 
                                     edgecolor='red', linewidth=2)
            ax.add_patch(rectangle)
            shapes.append('ÈïøÊñπÂΩ¢')
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.suptitle('Êï∞‰∏ÄÊï∞ÊúâÂ§öÂ∞ë‰∏™ÂõæÂΩ¢', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # ÁªüËÆ°ÂõæÂΩ¢Êï∞Èáè
    from collections import Counter
    shape_counts = Counter(shapes)
    
    print("\nÂõæÂΩ¢ÁªüËÆ°Ôºö")
    print("-" * 30)
    for shape, count in shape_counts.items():
        print(f"{shape}: {count}‰∏™")
    print("-" * 30)
    print(f"ÊÄªËÆ°: {len(shapes)}‰∏™")
    
    # ÊÄùËÄÉÈ¢ò
    print("\nü§î ÊÄùËÄÉÈ¢òÔºö")
    print(f"1. ÊúÄÂ§öÁöÑÊòØ‰ªÄ‰πàÂõæÂΩ¢Ôºü")
    most_common = shape_counts.most_common(1)[0]
    print(f"   Á≠îÊ°àÔºö{most_common[0]}ÔºåÊúâ{most_common[1]}‰∏™")
    print(f"2. ÊúÄÂ∞ëÁöÑÊòØ‰ªÄ‰πàÂõæÂΩ¢Ôºü")
    least_common = shape_counts.most_common()[-1]
    print(f"   Á≠îÊ°àÔºö{least_common[0]}ÔºåÊúâ{least_common[1]}‰∏™")

# ËøêË°åÂõæÂΩ¢ËÆ°Êï∞
count_shapes()
```
