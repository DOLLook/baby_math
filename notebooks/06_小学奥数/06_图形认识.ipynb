## 7. æ‰¾å›¾å½¢æ¸¸æˆ

æ¥ç©æ‰¾å›¾å½¢çš„æ¸¸æˆå§ï¼

```python
import random

def find_shape_game():
    """æ‰¾å›¾å½¢æ¸¸æˆ"""
    print("ğŸ® æ‰¾å›¾å½¢æ¸¸æˆ")
    print("=" * 40)
    print("è§„åˆ™ï¼šçœ‹å›¾å½¢ï¼Œè¯´å‡ºå®ƒçš„åå­—")
    
    score = 0
    total_questions = 5
    
    shapes = [
        {'name': 'åœ†å½¢', 'draw': lambda ax: ax.add_patch(plt.Circle((0, 0), 0.8, facecolor='lightblue', edgecolor='blue', linewidth=4))},
        {'name': 'æ­£æ–¹å½¢', 'draw': lambda ax: ax.add_patch(plt.Rectangle((-0.4, -0.4), 0.8, 0.8, facecolor='lightgreen', edgecolor='green', linewidth=4))},
        {'name': 'ä¸‰è§’å½¢', 'draw': lambda ax: ax.add_patch(plt.Polygon([(-0.4, -0.4), (0.4, -0.4), (0, 0.4)], facecolor='lightyellow', edgecolor='orange', linewidth=4))},
        {'name': 'é•¿æ–¹å½¢', 'draw': lambda ax: ax.add_patch(plt.Rectangle((-0.6, -0.4), 1.2, 0.8, facecolor='lightpink', edgecolor='red', linewidth=4)))}
    ]
    
    for i in range(total_questions):
        print(f"\né—®é¢˜ {i+1}/{total_questions}")
        
        # éšæœºé€‰æ‹©ä¸€ä¸ªå›¾å½¢
        shape = random.choice(shapes)
        
        # ç»˜åˆ¶å›¾å½¢
        fig, ax = plt.subplots(figsize=(6, 6))
        shape['draw'](ax)
        ax.set_xlim(-1.2, 1.2)
        ax.set_ylim(-1.2, 1.2)
        ax.set_aspect('equal')
        ax.axis('off')
        plt.tight_layout()
        plt.show()
        
        # è¯¢é—®ç­”æ¡ˆ
        answer = input("è¿™æ˜¯ä»€ä¹ˆå›¾å½¢ï¼Ÿ")
        
        if answer == shape['name']:
            print("âœ… ç­”å¯¹äº†ï¼")
            score += 1
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {shape['name']}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{total_questions}")
    
    if score == total_questions:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= total_questions * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
find_shape_game()
```

```python
def count_shapes():
    """å›¾å½¢è®¡æ•°"""
    print("ğŸ”¢ å›¾å½¢è®¡æ•°")
    print("=" * 40)
    
    # åˆ›å»ºéšæœºå›¾å½¢
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # ç”Ÿæˆéšæœºå›¾å½¢
    shapes = []
    for _ in range(10):
        shape_type = random.choice(['circle', 'square', 'triangle', 'rectangle'])
        x = random.uniform(-3, 3)
        y = random.uniform(-3, 3)
        
        if shape_type == 'circle':
            circle = plt.Circle((x, y), 0.3, facecolor='lightblue', 
                               edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
            shapes.append('åœ†å½¢')
        elif shape_type == 'square':
            square = plt.Rectangle((x-0.25, y-0.25), 0.5, 0.5, facecolor='lightgreen', 
                                  edgecolor='green', linewidth=2)
            ax.add_patch(square)
            shapes.append('æ­£æ–¹å½¢')
        elif shape_type == 'triangle':
            triangle = plt.Polygon([(x-0.3, y-0.25), (x+0.3, y-0.25), (x, y+0.25)], 
                                  facecolor='lightyellow', edgecolor='orange', linewidth=2)
            ax.add_patch(triangle)
            shapes.append('ä¸‰è§’å½¢')
        elif shape_type == 'rectangle':
            rectangle = plt.Rectangle((x-0.4, y-0.25), 0.8, 0.5, facecolor='lightpink', 
                                     edgecolor='red', linewidth=2)
            ax.add_patch(rectangle)
            shapes.append('é•¿æ–¹å½¢')
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.suptitle('æ•°ä¸€æ•°æœ‰å¤šå°‘ä¸ªå›¾å½¢', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # ç»Ÿè®¡å›¾å½¢æ•°é‡
    from collections import Counter
    shape_counts = Counter(shapes)
    
    print("\nå›¾å½¢ç»Ÿè®¡ï¼š")
    print("-" * 30)
    for shape, count in shape_counts.items():
        print(f"{shape}: {count}ä¸ª")
    print("-" * 30)
    print(f"æ€»è®¡: {len(shapes)}ä¸ª")
    
    # æ€è€ƒé¢˜
    print("\nğŸ¤” æ€è€ƒé¢˜ï¼š")
    print(f"1. æœ€å¤šçš„æ˜¯ä»€ä¹ˆå›¾å½¢ï¼Ÿ")
    most_common = shape_counts.most_common(1)[0]
    print(f"   ç­”æ¡ˆï¼š{most_common[0]}ï¼Œæœ‰{most_common[1]}ä¸ª")
    print(f"2. æœ€å°‘çš„æ˜¯ä»€ä¹ˆå›¾å½¢ï¼Ÿ")
    least_common = shape_counts.most_common()[-1]
    print(f"   ç­”æ¡ˆï¼š{least_common[0]}ï¼Œæœ‰{least_common[1]}ä¸ª")

# è¿è¡Œå›¾å½¢è®¡æ•°
count_shapes()
```
