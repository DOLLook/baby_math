## 1. è®¤è¯†åŠ æ³•

åŠ æ³•å°±æ˜¯æŠŠä¸¤ä¸ªæ•°åˆåœ¨ä¸€èµ·ï¼

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'SimSun']
plt.rcParams['axes.unicode_minus'] = False

def visualize_addition(a, b):
    """å¯è§†åŒ–åŠ æ³•"""
    fig, ax = plt.subplots(figsize=(12, 4))
    
    # ç»˜åˆ¶ç¬¬ä¸€ä¸ªæ•°
    for i in range(a):
        ax.add_patch(plt.Circle((i*1.2 + 0.5, 0.5), 0.4, 
                               facecolor='lightblue', edgecolor='blue', linewidth=2))
        ax.text(i*1.2 + 0.5, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # åŠ å·
    ax.text(a*1.2 + 1.0, 0.5, '+', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='green')
    
    # ç»˜åˆ¶ç¬¬äºŒä¸ªæ•°
    for i in range(b):
        ax.add_patch(plt.Circle((a*1.2 + 2.0 + i*1.2, 0.5), 0.4, 
                               facecolor='lightgreen', edgecolor='green', linewidth=2))
        ax.text(a*1.2 + 2.0 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # ç­‰å·
    ax.text((a+b)*1.2 + 2.5, 0.5, '=', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='red')
    
    # ç»“æœ
    result = a + b
    for i in range(result):
        ax.add_patch(plt.Circle((a+b)*1.2 + 3.5 + i*1.2, 0.5), 0.4, 
                               facecolor='lightyellow', edgecolor='orange', linewidth=2))
        ax.text((a+b)*1.2 + 3.5 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    ax.set_xlim(0, (a+b)*1.2 + 5.0)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(f'{a} + {b} = {result}', fontsize=20, fontweight='bold')
    plt.tight_layout()
    plt.show()

# å±•ç¤ºåŠ æ³•
print("â• åŠ æ³•æ¼”ç¤ºï¼š")
visualize_addition(3, 4)
visualize_addition(5, 2)
```

## 2. è®¤è¯†å‡æ³•

å‡æ³•å°±æ˜¯ä»ä¸€ä¸ªæ•°é‡Œæ‹¿èµ°ä¸€éƒ¨åˆ†ï¼

```python
def visualize_subtraction(a, b):
    """å¯è§†åŒ–å‡æ³•"""
    fig, ax = plt.subplots(figsize=(12, 4))
    
    # ç»˜åˆ¶è¢«å‡æ•°ï¼ˆå…¨éƒ¨ç”¨è“è‰²ï¼‰
    for i in range(a):
        ax.add_patch(plt.Circle((i*1.2 + 0.5, 0.5), 0.4, 
                               facecolor='lightblue', edgecolor='blue', linewidth=2))
        ax.text(i*1.2 + 0.5, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # å‡å·
    ax.text(a*1.2 + 1.0, 0.5, '-', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='red')
    
    # ç»˜åˆ¶å‡æ•°ï¼ˆç”¨çº¢è‰²è¡¨ç¤ºè¦æ‹¿èµ°çš„ï¼‰
    for i in range(b):
        ax.add_patch(plt.Circle((a*1.2 + 2.0 + i*1.2, 0.5), 0.4, 
                               facecolor='lightcoral', edgecolor='red', linewidth=2))
        ax.text(a*1.2 + 2.0 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # ç­‰å·
    ax.text((a+b)*1.2 + 2.5, 0.5, '=', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='green')
    
    # ç»“æœï¼ˆå‰©ä¸‹çš„ï¼‰
    result = a - b
    for i in range(result):
        ax.add_patch(plt.Circle((a+b)*1.2 + 3.5 + i*1.2, 0.5), 0.4, 
                               facecolor='lightgreen', edgecolor='green', linewidth=2))
        ax.text((a+b)*1.2 + 3.5 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    ax.set_xlim(0, (a+b)*1.2 + 5.0)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(f'{a} - {b} = {result}', fontsize=20, fontweight='bold')
    plt.tight_layout()
    plt.show()

# å±•ç¤ºå‡æ³•
print("â– å‡æ³•æ¼”ç¤ºï¼š")
visualize_subtraction(7, 3)
visualize_subtraction(9, 4)
```

## 3. åŠ æ³•è¡¨

è®©æˆ‘ä»¬çœ‹çœ‹10ä»¥å†…çš„åŠ æ³•è¡¨ï¼

```python
def create_addition_table():
    """åˆ›å»ºåŠ æ³•è¡¨"""
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # åˆ›å»ºåŠ æ³•è¡¨
    size = 11
    table_data = [[i + j for j in range(size)] for i in range(size)]
    
    # ç»˜åˆ¶è¡¨æ ¼
    for i in range(size):
        for j in range(size):
            # è®¾ç½®é¢œè‰²
            if i == 0 or j == 0:
                color = 'lightgray'
            elif i + j <= 10:
                color = 'lightblue'
            else:
                color = 'lightyellow'
            
            # ç»˜åˆ¶å•å…ƒæ ¼
            rect = plt.Rectangle((j, size-1-i), 1, 1, 
                                 facecolor=color, edgecolor='black', linewidth=1)
            ax.add_patch(rect)
            
            # æ·»åŠ æ•°å­—
            if i == 0 and j == 0:
                text = '+'
            elif i == 0:
                text = str(j)
            elif j == 0:
                text = str(i)
            else:
                text = str(table_data[i][j])
            
            ax.text(j + 0.5, size - 0.5 - i, text, 
                   fontsize=14, ha='center', va='center', fontweight='bold')
    
    ax.set_xlim(0, size)
    ax.set_ylim(0, size)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('10ä»¥å†…åŠ æ³•è¡¨', fontsize=18, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

# æ˜¾ç¤ºåŠ æ³•è¡¨
create_addition_table()
```

## 4. å‡‘åæ³•

å­¦ä¹ å‡‘åæ³•ï¼Œè®©åŠ æ³•è®¡ç®—æ›´ç®€å•ï¼

```python
def make_ten_method(a, b):
    """å‡‘åæ³•æ¼”ç¤º"""
    if a + b > 10:
        print("è¿™ä¸ªå’Œå¤§äº10ï¼Œæˆ‘ä»¬ä¸‹æ¬¡å­¦ä¹ è¿›ä½åŠ æ³•ï¼")
        return
    
    fig, ax = plt.subplots(figsize=(14, 4))
    
    # æ‰¾å‡ºå“ªä¸ªæ•°æ›´æ¥è¿‘10
    if a >= b:
        first, second = a, b
        color1, color2 = 'lightblue', 'lightgreen'
    else:
        first, second = b, a
        color1, color2 = 'lightgreen', 'lightblue'
    
    # è®¡ç®—éœ€è¦å‡‘åçš„æ•°é‡
    needed = 10 - first
    
    # ç»˜åˆ¶ç¬¬ä¸€ä¸ªæ•°
    for i in range(first):
        ax.add_patch(plt.Circle((i*1.2 + 0.5, 0.5), 0.4, 
                               facecolor=color1, edgecolor='blue', linewidth=2))
        ax.text(i*1.2 + 0.5, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # åŠ å·
    ax.text(first*1.2 + 1.0, 0.5, '+', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='green')
    
    # ç»˜åˆ¶ç¬¬äºŒä¸ªæ•°
    for i in range(second):
        if i < needed:
            # å‡‘åçš„éƒ¨åˆ†ç”¨ç‰¹æ®Šé¢œè‰²
            ax.add_patch(plt.Circle((first*1.2 + 2.0 + i*1.2, 0.5), 0.4, 
                                   facecolor='yellow', edgecolor='orange', linewidth=3))
        else:
            ax.add_patch(plt.Circle((first*1.2 + 2.0 + i*1.2, 0.5), 0.4, 
                                   facecolor=color2, edgecolor='green', linewidth=2))
        ax.text(first*1.2 + 2.0 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    # ç­‰å·
    ax.text((first+second)*1.2 + 2.5, 0.5, '=', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='red')
    
    # ç»“æœ
    result = first + second
    for i in range(result):
        ax.add_patch(plt.Circle((first+second)*1.2 + 3.5 + i*1.2, 0.5), 0.4, 
                               facecolor='lightpink', edgecolor='red', linewidth=2))
        ax.text((first+second)*1.2 + 3.5 + i*1.2, 0.5, str(i+1), fontsize=12, ha='center', va='center', fontweight='bold')
    
    ax.set_xlim(0, (first+second)*1.2 + 5.0)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    
    title = f'{a} + {b} = {result} (å‡‘åæ³•: {first} + {needed} = 10, 10 + {second-needed} = {result})'
    ax.set_title(title, fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# å‡‘åæ³•æ¼”ç¤º
print("ğŸ”Ÿ å‡‘åæ³•æ¼”ç¤ºï¼š")
make_ten_method(8, 3)
make_ten_method(7, 5)
make_ten_method(9, 2)
```

## 5. æ•°å­¦æ•…äº‹ï¼šå°å…”å­æ‹”èåœ

é€šè¿‡æ•…äº‹æ¥å­¦ä¹ åŠ å‡æ³•ï¼

```python
def rabbit_carrot_story():
    """å°å…”å­æ‹”èåœçš„æ•…äº‹"""
    import random
    
    print("ğŸ° å°å…”å­æ‹”èåœçš„æ•…äº‹")
    print("=" * 40)
    
    # ç¬¬ä¸€å¤©
    morning = random.randint(3, 6)
    afternoon = random.randint(2, 4)
    total = morning + afternoon
    
    print(f"\nç¬¬ä¸€å¤©æ—©ä¸Šï¼Œå°å…”å­æ‹”äº† {morning} ä¸ªèåœ")
    print(f"ç¬¬ä¸€å¤©ä¸‹åˆï¼Œå°å…”å­åˆæ‹”äº† {afternoon} ä¸ªèåœ")
    print(f"å°å…”å­ç¬¬ä¸€å¤©ä¸€å…±æ‹”äº†å‡ ä¸ªèåœï¼Ÿ")
    print(f"ç­”æ¡ˆï¼š{morning} + {afternoon} = {total} ä¸ªèåœ")
    
    # ç¬¬äºŒå¤©
    print(f"\nç¬¬äºŒå¤©ï¼Œå°å…”å­æ‹”äº† {total + 2} ä¸ªèåœ")
    eaten = random.randint(1, 3)
    left = (total + 2) - eaten
    
    print(f"å°å…”å­åƒäº† {eaten} ä¸ªèåœ")
    print(f"è¿˜å‰©ä¸‹å‡ ä¸ªèåœï¼Ÿ")
    print(f"ç­”æ¡ˆï¼š{total + 2} - {eaten} = {left} ä¸ªèåœ")
    
    # å¯è§†åŒ–
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # ç¬¬ä¸€å¤©
    for i in range(morning):
        ax1.add_patch(plt.Circle((0.5, 0.8 - i*0.15), 0.08, 
                               facecolor='orange', edgecolor='brown', linewidth=2))
    for i in range(afternoon):
        ax1.add_patch(plt.Circle((0.8, 0.8 - i*0.15), 0.08, 
                               facecolor='orange', edgecolor='brown', linewidth=2))
    
    ax1.text(0.5, 0.2, f'æ—©ä¸Š: {morning}', fontsize=12, ha='center', fontweight='bold')
    ax1.text(0.8, 0.2, f'ä¸‹åˆ: {afternoon}', fontsize=12, ha='center', fontweight='bold')
    ax1.text(0.65, 0.1, f'æ€»è®¡: {total}', fontsize=14, ha='center', 
            fontweight='bold', color='red')
    ax1.set_xlim(0, 1.3)
    ax1.set_ylim(0, 1)
    ax1.set_aspect('equal')
    ax1.axis('off')
    ax1.set_title('ç¬¬ä¸€å¤©ï¼šæ‹”èåœ', fontsize=14, fontweight='bold')
    
    # ç¬¬äºŒå¤©
    for i in range(total + 2):
        if i < eaten:
            color = 'lightcoral'  # è¢«åƒæ‰çš„
        else:
            color = 'orange'  # å‰©ä¸‹çš„
        ax2.add_patch(plt.Circle((0.3 + (i % 5)*0.15, 0.8 - (i // 5)*0.15), 0.06, 
                               facecolor=color, edgecolor='brown', linewidth=1))
    
    ax2.text(0.5, 0.2, f'æ‹”äº†: {total + 2}', fontsize=12, ha='center', fontweight='bold')
    ax2.text(0.3, 0.1, f'åƒäº†: {eaten}', fontsize=12, ha='center', 
            fontweight='bold', color='red')
    ax2.text(0.7, 0.1, f'å‰©ä¸‹: {left}', fontsize=12, ha='center', 
            fontweight='bold', color='green')
    ax2.set_xlim(0, 1)
    ax2.set_ylim(0, 1)
    ax2.set_aspect('equal')
    ax2.axis('off')
    ax2.set_title('ç¬¬äºŒå¤©ï¼šåƒèåœ', fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# è¿è¡Œæ•…äº‹
rabbit_carrot_story()
```

## 6. åŠ å‡æ³•æ¸¸æˆ

æ¥ç©ä¸€ä¸ªæœ‰è¶£çš„åŠ å‡æ³•æ¸¸æˆå§ï¼

```python
import random

def addition_subtraction_game():
    """åŠ å‡æ³•æ¸¸æˆ"""
    print("ğŸ® åŠ å‡æ³•å¤§æŒ‘æˆ˜")
    print("=" * 40)
    
    score = 0
    total_questions = 5
    
    for i in range(total_questions):
        print(f"\né—®é¢˜ {i+1}/{total_questions}")
        
        # éšæœºé€‰æ‹©åŠ æ³•æˆ–å‡æ³•
        if random.choice([True, False]):
            # åŠ æ³•
            a = random.randint(1, 9)
            b = random.randint(1, 9)
            if a + b > 10:
                continue
            question = f"{a} + {b} = ?"
            answer = a + b
        else:
            # å‡æ³•
            a = random.randint(3, 10)
            b = random.randint(1, a-1)
            question = f"{a} - {b} = ?"
            answer = a - b
        
        print(question)
        user_answer = input("ä½ çš„ç­”æ¡ˆï¼š")
        
        try:
            if int(user_answer) == answer:
                print("âœ… ç­”å¯¹äº†ï¼å¤ªæ£’äº†ï¼")
                score += 1
            else:
                print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {answer}")
        except:
            print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {answer}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{total_questions}")
    
    if score == total_questions:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= total_questions * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
addition_subtraction_game()
```

## 7. æ‰¾è§„å¾‹å¡«ç©º

æ‰¾å‡ºæ•°å­—åºåˆ—çš„è§„å¾‹ï¼Œå¡«ä¸Šæ­£ç¡®çš„æ•°å­—ï¼

```python
def find_pattern():
    """æ‰¾è§„å¾‹å¡«ç©º"""
    patterns = [
        {
            'sequence': [1, 2, 3, 4, 5, ?],
            'rule': 'æ¯æ¬¡åŠ 1',
            'answer': 6
        },
        {
            'sequence': [2, 4, 6, 8, 10, ?],
            'rule': 'æ¯æ¬¡åŠ 2',
            'answer': 12
        },
        {
            'sequence': [10, 9, 8, 7, 6, ?],
            'rule': 'æ¯æ¬¡å‡1',
            'answer': 5
        },
        {
            'sequence': [1, 3, 5, 7, 9, ?],
            'rule': 'æ¯æ¬¡åŠ 2ï¼ˆå¥‡æ•°ï¼‰',
            'answer': 11
        },
        {
            'sequence': [5, 4, 3, 2, 1, ?],
            'rule': 'æ¯æ¬¡å‡1',
            'answer': 0
        }
    ]
    
    print("ğŸ” æ‰¾è§„å¾‹å¡«ç©º")
    print("=" * 40)
    
    score = 0
    
    for i, pattern in enumerate(patterns):
        print(f"\né—®é¢˜ {i+1}:")
        seq_str = ', '.join([str(x) if x != '?' else '?' for x in pattern['sequence']])
        print(f"åºåˆ—ï¼š{seq_str}")
        print(f"æç¤ºï¼š{pattern['rule']}")
        
        answer = input("é—®å·å¤„åº”è¯¥å¡«ä»€ä¹ˆæ•°å­—ï¼Ÿ")
        
        try:
            if int(answer) == pattern['answer']:
                print("âœ… ç­”å¯¹äº†ï¼")
                score += 1
            else:
                print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {pattern['answer']}")
        except:
            print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {pattern['answer']}")
    
    print(f"\nğŸ¯ ä½ çš„å¾—åˆ†ï¼š{score}/{len(patterns)}")

# è¿è¡Œæ‰¾è§„å¾‹
find_pattern()
```

## 8. åŠ å‡æ³•å…³ç³»å›¾

çœ‹çœ‹åŠ æ³•å’Œå‡æ³•çš„å…³ç³»ï¼

```python
def addition_subtraction_relation():
    """åŠ å‡æ³•å…³ç³»å›¾"""
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # ç¤ºä¾‹ï¼š3 + 5 = 8, 8 - 3 = 5, 8 - 5 = 3
    examples = [
        (2, 3, 5),
        (3, 4, 7),
        (4, 5, 9),
        (1, 6, 7),
        (5, 5, 10)
    ]
    
    positions = [
        (0.2, 0.8),
        (0.5, 0.8),
        (0.8, 0.8),
        (0.2, 0.5),
        (0.5, 0.5),
        (0.8, 0.5),
        (0.2, 0.2),
        (0.5, 0.2),
        (0.8, 0.2)
    ]
    
    for i, (a, b, c) in enumerate(examples):
        x, y = positions[i]
        
        # ç»˜åˆ¶ä¸‰è§’å½¢å…³ç³»
        # åŠ æ•°1
        ax.add_patch(plt.Circle((x, y), 0.08, facecolor='lightblue', edgecolor='blue', linewidth=2))
        ax.text(x, y, str(a), fontsize=14, ha='center', va='center', fontweight='bold')
        
        # åŠ æ•°2
        ax.add_patch(plt.Circle((x + 0.15, y), 0.08, facecolor='lightgreen', edgecolor='green', linewidth=2))
        ax.text(x + 0.15, y, str(b), fontsize=14, ha='center', va='center', fontweight='bold')
        
        # å’Œ
        ax.add_patch(plt.Circle((x + 0.075, y - 0.12), 0.08, facecolor='lightyellow', edgecolor='orange', linewidth=2))
        ax.text(x + 0.075, y - 0.12, str(c), fontsize=14, ha='center', va='center', fontweight='bold')
        
        # åŠ å·
        ax.text(x + 0.075, y + 0.05, '+', fontsize=16, ha='center', va='center', fontweight='bold', color='green')
        
        # ç­‰å·
        ax.text(x + 0.075, y - 0.06, '=', fontsize=16, ha='center', va='center', fontweight='bold', color='red')
        
        # å‡æ³•å…³ç³»
        ax.text(x + 0.075, y - 0.18, f'{c}-{a}={b}, {c}-{b}={a}', 
                fontsize=8, ha='center', va='center', color='purple')
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('åŠ æ³•å’Œå‡æ³•çš„å…³ç³»', fontsize=18, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

# æ˜¾ç¤ºå…³ç³»å›¾
addition_subtraction_relation()
```
