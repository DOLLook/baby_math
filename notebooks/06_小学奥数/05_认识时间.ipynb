## 1. 认识钟面

让我们认识钟面的各个部分！

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'SimSun']
plt.rcParams['axes.unicode_minus'] = False

def draw_clock_face():
    """绘制钟面"""
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # 绘制钟面外圈
    clock_face = plt.Circle((0, 0), 1, facecolor='white', 
                           edgecolor='black', linewidth=4)
    ax.add_patch(clock_face)
    
    # 绘制刻度
    for i in range(12):
        angle = i * np.pi / 6
        x1 = 0.9 * np.cos(angle)
        y1 = 0.9 * np.sin(angle)
        x2 = 0.95 * np.cos(angle)
        y2 = 0.95 * np.sin(angle)
        ax.plot([x1, x2], [y1, y2], 'black', linewidth=3)
        
        # 添加数字
        num_x = 0.75 * np.cos(angle - np.pi/2)
        num_y = 0.75 * np.sin(angle - np.pi/2)
        num = 12 if i == 0 else i
        ax.text(num_x, num_y, str(num), fontsize=16, ha='center', va='center', fontweight='bold')
    
    # 绘制时针（指向12）
    hour_x = 0.5 * np.cos(-np.pi/2)
    hour_y = 0.5 * np.sin(-np.pi/2)
    ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=6, label='时针')
    
    # 绘制分针（指向12）
    minute_x = 0.7 * np.cos(-np.pi/2)
    minute_y = 0.7 * np.sin(-np.pi/2)
    ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=4, label='分针')
    
    # 绘制中心点
    center = plt.Circle((0, 0), 0.05, facecolor='black', edgecolor='black')
    ax.add_patch(center)
    
    # 标注
    ax.text(0, -1.3, '时针（短针）', fontsize=14, ha='center', color='red', fontweight='bold')
    ax.text(0, -1.45, '分针（长针）', fontsize=14, ha='center', color='blue', fontweight='bold')
    ax.text(0, 1.2, '钟面', fontsize=18, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# 显示钟面
print("⏰ 认识钟面：")
print("钟面有12个数字，从1到12")
print("时针是短针，走得慢")
print("分针是长针，走得快")
draw_clock_face()
```

## 2. 认识整点

当分针指向12时，就是整点！

```python
def draw_clock(hour, minute, title):
    """绘制指定时间的钟表"""
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # 绘制钟面外圈
    clock_face = plt.Circle((0, 0), 1, facecolor='white', 
                           edgecolor='black', linewidth=4)
    ax.add_patch(clock_face)
    
    # 绘制刻度
    for i in range(12):
        angle = i * np.pi / 6
        x1 = 0.9 * np.cos(angle)
        y1 = 0.9 * np.sin(angle)
        x2 = 0.95 * np.cos(angle)
        y2 = 0.95 * np.sin(angle)
        ax.plot([x1, x2], [y1, y2], 'black', linewidth=3)
        
        # 添加数字
        num_x = 0.75 * np.cos(angle - np.pi/2)
        num_y = 0.75 * np.sin(angle - np.pi/2)
        num = 12 if i == 0 else i
        ax.text(num_x, num_y, str(num), fontsize=16, ha='center', va='center', fontweight='bold')
    
    # 计算时针角度
    hour_angle = (hour % 12 + minute / 60) * np.pi / 6 - np.pi/2
    hour_x = 0.5 * np.cos(hour_angle)
    hour_y = 0.5 * np.sin(hour_angle)
    ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=6)
    
    # 计算分针角度
    minute_angle = minute * np.pi / 30 - np.pi/2
    minute_x = 0.7 * np.cos(minute_angle)
    minute_y = 0.7 * np.sin(minute_angle)
    ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=4)
    
    # 绘制中心点
    center = plt.Circle((0, 0), 0.05, facecolor='black', edgecolor='black')
    ax.add_patch(center)
    
    # 标题
    ax.text(0, 1.3, title, fontsize=18, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# 展示整点
print("🕐 认识整点：")
print("当分针指向12时，就是整点")
print("时针指向几，就是几点")

draw_clock(1, 0, '1点整')
draw_clock(3, 0, '3点整')
draw_clock(6, 0, '6点整')
draw_clock(9, 0, '9点整')
draw_clock(12, 0, '12点整')
```

## 3. 认识半点

当分针指向6时，就是半点！

```python
# 展示半点
print("🕑 认识半点：")
print("当分针指向6时，就是半点")
print("时针指向两个数字中间，就是几点半")

draw_clock(1, 30, '1点半')
draw_clock(3, 30, '3点半')
draw_clock(6, 30, '6点半')
draw_clock(9, 30, '9点半')
draw_clock(12, 30, '12点半')
```

## 4. 时间排序

学习按时间的顺序排列！

```python
def sort_times():
    """时间排序"""
    print("📊 时间排序")
    print("=" * 40)
    
    # 创建一些时间
    times = [
        {'hour': 3, 'minute': 0, 'label': '3点'},
        {'hour': 9, 'minute': 0, 'label': '9点'},
        {'hour': 6, 'minute': 30, 'label': '6点半'},
        {'hour': 1, 'minute': 30, 'label': '1点半'},
        {'hour': 12, 'minute': 0, 'label': '12点'}
    ]
    
    print("\n原始顺序：")
    for i, time in enumerate(times):
        print(f"{i+1}. {time['label']}")
    
    # 按时间排序
    sorted_times = sorted(times, key=lambda x: x['hour'] * 60 + x['minute'])
    
    print("\n按时间从早到晚排序：")
    for i, time in enumerate(sorted_times):
        print(f"{i+1}. {time['label']}")
    
    # 可视化
    fig, axes = plt.subplots(1, 5, figsize=(20, 4))
    
    for i, time in enumerate(sorted_times):
        ax = axes[i]
        
        # 绘制钟面外圈
        clock_face = plt.Circle((0, 0), 0.9, facecolor='white', 
                               edgecolor='black', linewidth=3)
        ax.add_patch(clock_face)
        
        # 绘制刻度
        for j in range(12):
            angle = j * np.pi / 6
            x1 = 0.8 * np.cos(angle)
            y1 = 0.8 * np.sin(angle)
            x2 = 0.85 * np.cos(angle)
            y2 = 0.85 * np.sin(angle)
            ax.plot([x1, x2], [y1, y2], 'black', linewidth=2)
            
            # 添加数字
            num_x = 0.65 * np.cos(angle - np.pi/2)
            num_y = 0.65 * np.sin(angle - np.pi/2)
            num = 12 if j == 0 else j
            ax.text(num_x, num_y, str(num), fontsize=10, ha='center', va='center', fontweight='bold')
        
        # 计算时针角度
        hour_angle = (time['hour'] % 12 + time['minute'] / 60) * np.pi / 6 - np.pi/2
        hour_x = 0.45 * np.cos(hour_angle)
        hour_y = 0.45 * np.sin(hour_angle)
        ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=5)
        
        # 计算分针角度
        minute_angle = time['minute'] * np.pi / 30 - np.pi/2
        minute_x = 0.65 * np.cos(minute_angle)
        minute_y = 0.65 * np.sin(minute_angle)
        ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=3)
        
        # 绘制中心点
        center = plt.Circle((0, 0), 0.04, facecolor='black', edgecolor='black')
        ax.add_patch(center)
        
        # 标题
        ax.text(0, -1.2, f"第{i+1}名", fontsize=12, ha='center', fontweight='bold')
        ax.text(0, 1.2, time['label'], fontsize=12, ha='center', fontweight='bold')
        
        ax.set_xlim(-1.3, 1.3)
        ax.set_ylim(-1.3, 1.3)
        ax.set_aspect('equal')
        ax.axis('off')
    
    plt.suptitle('按时间从早到晚排序', fontsize=16, fontweight='bold', y=1.05)
    plt.tight_layout()
    plt.show()

# 运行时间排序
sort_times()
```

## 5. 数学故事：小明的一天

通过故事学习时间！

```python
def xiaoming_day():
    """小明的一天"""
    print("📖 小明的一天")
    print("=" * 40)
    
    schedule = [
        {'time': '7点', 'activity': '起床', 'hour': 7, 'minute': 0},
        {'time': '7点半', 'activity': '吃早饭', 'hour': 7, 'minute': 30},
        {'time': '8点', 'activity': '上学', 'hour': 8, 'minute': 0},
        {'time': '12点', 'activity': '吃午饭', 'hour': 12, 'minute': 0},
        {'time': '4点', 'activity': '放学回家', 'hour': 4, 'minute': 0},
        {'time': '6点', 'activity': '吃晚饭', 'hour': 6, 'minute': 0},
        {'time': '9点', 'activity': '睡觉', 'hour': 9, 'minute': 0}
    ]
    
    print("\n小明一天的时间表：")
    for i, event in enumerate(schedule):
        print(f"{i+1}. {event['time']} - {event['activity']}")
    
    # 可视化
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    axes = axes.flatten()
    
    for i, event in enumerate(schedule):
        ax = axes[i]
        
        # 绘制钟面外圈
        clock_face = plt.Circle((0, 0), 0.8, facecolor='white', 
                               edgecolor='black', linewidth=3)
        ax.add_patch(clock_face)
        
        # 绘制刻度
        for j in range(12):
            angle = j * np.pi / 6
            x1 = 0.7 * np.cos(angle)
            y1 = 0.7 * np.sin(angle)
            x2 = 0.75 * np.cos(angle)
            y2 = 0.75 * np.sin(angle)
            ax.plot([x1, x2], [y1, y2], 'black', linewidth=2)
            
            # 添加数字
            num_x = 0.55 * np.cos(angle - np.pi/2)
            num_y = 0.55 * np.sin(angle - np.pi/2)
            num = 12 if j == 0 else j
            ax.text(num_x, num_y, str(num), fontsize=8, ha='center', va='center', fontweight='bold')
        
        # 计算时针角度
        hour_angle = (event['hour'] % 12 + event['minute'] / 60) * np.pi / 6 - np.pi/2
        hour_x = 0.4 * np.cos(hour_angle)
        hour_y = 0.4 * np.sin(hour_angle)
        ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=4)
        
        # 计算分针角度
        minute_angle = event['minute'] * np.pi / 30 - np.pi/2
        minute_x = 0.6 * np.cos(minute_angle)
        minute_y = 0.6 * np.sin(minute_angle)
        ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=3)
        
        # 绘制中心点
        center = plt.Circle((0, 0), 0.03, facecolor='black', edgecolor='black')
        ax.add_patch(center)
        
        # 标题
        ax.text(0, -1.1, event['time'], fontsize=10, ha='center', fontweight='bold')
        ax.text(0, 1.1, event['activity'], fontsize=9, ha='center', fontweight='bold')
        
        ax.set_xlim(-1.2, 1.2)
        ax.set_ylim(-1.2, 1.2)
        ax.set_aspect('equal')
        ax.axis('off')
    
    # 隐藏最后一个子图
    axes[7].axis('off')
    
    plt.suptitle('小明的一天', fontsize=18, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()
    
    # 思考题
    print("\n🤔 思考题：")
    print("1. 小明几点起床？")
    print("   答案：7点")
    print("2. 小明几点吃午饭？")
    print("   答案：12点")
    print("3. 小明几点放学回家？")
    print("   答案：4点")
    print("4. 小明几点睡觉？")
    print("   答案：9点")

# 运行故事
xiaoming_day()
```

## 6. 认时间游戏

来玩认时间的游戏吧！

```python
import random

def time_game():
    """认时间游戏"""
    print("🎮 认时间游戏")
    print("=" * 40)
    print("规则：看钟表，说出是几点")
    
    score = 0
    total_questions = 5
    
    for i in range(total_questions):
        print(f"\n问题 {i+1}/{total_questions}")
        
        # 随机生成时间（整点或半点）
        hour = random.randint(1, 12)
        minute = random.choice([0, 30])
        
        # 绘制钟表
        fig, ax = plt.subplots(figsize=(6, 6))
        
        # 绘制钟面外圈
        clock_face = plt.Circle((0, 0), 1, facecolor='white', 
                               edgecolor='black', linewidth=4)
        ax.add_patch(clock_face)
        
        # 绘制刻度
        for j in range(12):
            angle = j * np.pi / 6
            x1 = 0.9 * np.cos(angle)
            y1 = 0.9 * np.sin(angle)
            x2 = 0.95 * np.cos(angle)
            y2 = 0.95 * np.sin(angle)
            ax.plot([x1, x2], [y1, y2], 'black', linewidth=3)
            
            # 添加数字
            num_x = 0.75 * np.cos(angle - np.pi/2)
            num_y = 0.75 * np.sin(angle - np.pi/2)
            num = 12 if j == 0 else j
            ax.text(num_x, num_y, str(num), fontsize=16, ha='center', va='center', fontweight='bold')
        
        # 计算时针角度
        hour_angle = (hour % 12 + minute / 60) * np.pi / 6 - np.pi/2
        hour_x = 0.5 * np.cos(hour_angle)
        hour_y = 0.5 * np.sin(hour_angle)
        ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=6)
        
        # 计算分针角度
        minute_angle = minute * np.pi / 30 - np.pi/2
        minute_x = 0.7 * np.cos(minute_angle)
        minute_y = 0.7 * np.sin(minute_angle)
        ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=4)
        
        # 绘制中心点
        center = plt.Circle((0, 0), 0.05, facecolor='black', edgecolor='black')
        ax.add_patch(center)
        
        ax.set_xlim(-1.5, 1.5)
        ax.set_ylim(-1.5, 1.5)
        ax.set_aspect('equal')
        ax.axis('off')
        plt.tight_layout()
        plt.show()
        
        # 询问答案
        if minute == 0:
            correct_answer = f"{hour}点"
        else:
            correct_answer = f"{hour}点半"
        
        answer = input("这是几点？")
        
        if answer == correct_answer:
            print("✅ 答对了！")
            score += 1
        else:
            print(f"❌ 不对哦，正确答案是 {correct_answer}")
    
    print(f"\n🎯 游戏结束！")
    print(f"你的得分：{score}/{total_questions}")
    
    if score == total_questions:
        print("🏆 太厉害了！全部正确！")
    elif score >= total_questions * 0.8:
        print("🌟 做得很好！继续保持！")
    else:
        print("💪 继续加油，多练习就会进步！")

# 运行游戏
time_game()
```

## 7. 时间长短比较

学习比较时间的长短！

```python
def compare_time_duration():
    """比较时间长短"""
    print("⏱️ 时间长短比较")
    print("=" * 40)
    
    comparisons = [
        {'time1': '1小时', 'time2': '30分钟', 'longer': '1小时', 'reason': '1小时 = 60分钟，比30分钟长'},
        {'time1': '1小时', 'time2': '2小时', 'longer': '2小时', 'reason': '2小时比1小时长'},
        {'time1': '30分钟', 'time2': '1小时', 'longer': '1小时', 'reason': '1小时 = 60分钟，比30分钟长'},
        {'time1': '1小时', 'time2': '1小时', 'longer': '一样长', 'reason': '都是1小时，时间相同'},
        {'time1': '2小时', 'time2': '1小时', 'longer': '2小时', 'reason': '2小时比1小时长'}
    ]
    
    score = 0
    
    for i, comp in enumerate(comparisons):
        print(f"\n问题 {i+1}:")
        print(f"{comp['time1']} 和 {comp['time2']}，哪个时间长？")
        
        answer = input("你的答案：")
        
        if answer == comp['longer']:
            print("✅ 答对了！")
            print(f"原因：{comp['reason']}")
            score += 1
        else:
            print(f"❌ 不对哦，正确答案是 {comp['longer']}")
            print(f"原因：{comp['reason']}")
    
    print(f"\n🎯 你的得分：{score}/{len(comparisons)}")
    
    if score == len(comparisons):
        print("🏆 太厉害了！全部正确！")
    elif score >= len(comparisons) * 0.8:
        print("🌟 做得很好！继续保持！")
    else:
        print("💪 继续加油，多练习就会进步！")

# 运行时间比较
compare_time_duration()
```
