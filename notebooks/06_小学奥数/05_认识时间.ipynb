```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'SimSun']
plt.rcParams['axes.unicode_minus'] = False

def draw_clock_face():
    """ç»˜åˆ¶é’Ÿé¢"""
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # ç»˜åˆ¶é’Ÿé¢å¤–åœˆ
    clock_face = plt.Circle((0, 0), 1, facecolor='white', 
                           edgecolor='black', linewidth=4)
    ax.add_patch(clock_face)
    
    # ç»˜åˆ¶åˆ»åº¦
    for i in range(12):
        angle = i * np.pi / 6
        x1 = 0.9 * np.cos(angle)
        y1 = 0.9 * np.sin(angle)
        x2 = 0.95 * np.cos(angle)
        y2 = 0.95 * np.sin(angle)
        ax.plot([x1, x2], [y1, y2], 'black', linewidth=3)
        
        # æ·»åŠ æ•°å­—
        num_x = 0.75 * np.cos(angle - np.pi/2)
        num_y = 0.75 * np.sin(angle - np.pi/2)
        num = 12 if i == 0 else i
        ax.text(num_x, num_y, str(num), fontsize=16, ha='center', va='center', fontweight='bold')
    
    # ç»˜åˆ¶æ—¶é’ˆï¼ˆæŒ‡å‘12ï¼‰
    hour_x = 0.5 * np.cos(-np.pi/2)
    hour_y = 0.5 * np.sin(-np.pi/2)
    ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=6, label='æ—¶é’ˆ')
    
    # ç»˜åˆ¶åˆ†é’ˆï¼ˆæŒ‡å‘12ï¼‰
    minute_x = 0.7 * np.cos(-np.pi/2)
    minute_y = 0.7 * np.sin(-np.pi/2)
    ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=4, label='åˆ†é’ˆ')
    
    # ç»˜åˆ¶ä¸­å¿ƒç‚¹
    center = plt.Circle((0, 0), 0.05, facecolor='black', edgecolor='black')
    ax.add_patch(center)
    
    # æ ‡æ³¨
    ax.text(0, -1.3, 'æ—¶é’ˆï¼ˆçŸ­é’ˆï¼‰', fontsize=14, ha='center', color='red', fontweight='bold')
    ax.text(0, -1.45, 'åˆ†é’ˆï¼ˆé•¿é’ˆï¼‰', fontsize=14, ha='center', color='blue', fontweight='bold')
    ax.text(0, 1.2, 'é’Ÿé¢', fontsize=18, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# æ˜¾ç¤ºé’Ÿé¢
print("â° è®¤è¯†é’Ÿé¢ï¼š")
print("é’Ÿé¢æœ‰12ä¸ªæ•°å­—ï¼Œä»1åˆ°12")
print("æ—¶é’ˆæ˜¯çŸ­é’ˆï¼Œèµ°å¾—æ…¢")
print("åˆ†é’ˆæ˜¯é•¿é’ˆï¼Œèµ°å¾—å¿«")
draw_clock_face()
```

```python
def draw_clock(hour, minute, title):
    """ç»˜åˆ¶æŒ‡å®šæ—¶é—´çš„é’Ÿè¡¨"""
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # ç»˜åˆ¶é’Ÿé¢å¤–åœˆ
    clock_face = plt.Circle((0, 0), 1, facecolor='white', 
                           edgecolor='black', linewidth=4)
    ax.add_patch(clock_face)
    
    # ç»˜åˆ¶åˆ»åº¦
    for i in range(12):
        angle = i * np.pi / 6
        x1 = 0.9 * np.cos(angle)
        y1 = 0.9 * np.sin(angle)
        x2 = 0.95 * np.cos(angle)
        y2 = 0.95 * np.sin(angle)
        ax.plot([x1, x2], [y1, y2], 'black', linewidth=3)
        
        # æ·»åŠ æ•°å­—
        num_x = 0.75 * np.cos(angle - np.pi/2)
        num_y = 0.75 * np.sin(angle - np.pi/2)
        num = 12 if i == 0 else i
        ax.text(num_x, num_y, str(num), fontsize=16, ha='center', va='center', fontweight='bold')
    
    # è®¡ç®—æ—¶é’ˆè§’åº¦
    hour_angle = (hour % 12 + minute / 60) * np.pi / 6 - np.pi/2
    hour_x = 0.5 * np.cos(hour_angle)
    hour_y = 0.5 * np.sin(hour_angle)
    ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=6)
    
    # è®¡ç®—åˆ†é’ˆè§’åº¦
    minute_angle = minute * np.pi / 30 - np.pi/2
    minute_x = 0.7 * np.cos(minute_angle)
    minute_y = 0.7 * np.sin(minute_angle)
    ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=4)
    
    # ç»˜åˆ¶ä¸­å¿ƒç‚¹
    center = plt.Circle((0, 0), 0.05, facecolor='black', edgecolor='black')
    ax.add_patch(center)
    
    # æ ‡é¢˜
    ax.text(0, 1.3, title, fontsize=18, ha='center', fontweight='bold')
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# å±•ç¤ºæ•´ç‚¹
print("ğŸ• è®¤è¯†æ•´ç‚¹ï¼š")
print("å½“åˆ†é’ˆæŒ‡å‘12æ—¶ï¼Œå°±æ˜¯æ•´ç‚¹")
print("æ—¶é’ˆæŒ‡å‘å‡ ï¼Œå°±æ˜¯å‡ ç‚¹")

draw_clock(1, 0, '1ç‚¹æ•´')
draw_clock(3, 0, '3ç‚¹æ•´')
draw_clock(6, 0, '6ç‚¹æ•´')
draw_clock(9, 0, '9ç‚¹æ•´')
draw_clock(12, 0, '12ç‚¹æ•´')
```

```python
def sort_times():
    """æ—¶é—´æ’åº"""
    print("ğŸ“Š æ—¶é—´æ’åº")
    print("=" * 40)
    
    # åˆ›å»ºä¸€äº›æ—¶é—´
    times = [
        {'hour': 3, 'minute': 0, 'label': '3ç‚¹'},
        {'hour': 9, 'minute': 0, 'label': '9ç‚¹'},
        {'hour': 6, 'minute': 30, 'label': '6ç‚¹åŠ'},
        {'hour': 1, 'minute': 30, 'label': '1ç‚¹åŠ'},
        {'hour': 12, 'minute': 0, 'label': '12ç‚¹'}
    ]
    
    print("\nåŸå§‹é¡ºåºï¼š")
    for i, time in enumerate(times):
        print(f"{i+1}. {time['label']}")
    
    # æŒ‰æ—¶é—´æ’åº
    sorted_times = sorted(times, key=lambda x: x['hour'] * 60 + x['minute'])
    
    print("\næŒ‰æ—¶é—´ä»æ—©åˆ°æ™šæ’åºï¼š")
    for i, time in enumerate(sorted_times):
        print(f"{i+1}. {time['label']}")
    
    # å¯è§†åŒ–
    fig, axes = plt.subplots(1, 5, figsize=(20, 4))
    
    for i, time in enumerate(sorted_times):
        ax = axes[i]
        
        # ç»˜åˆ¶é’Ÿé¢å¤–åœˆ
        clock_face = plt.Circle((0, 0), 0.9, facecolor='white', 
                               edgecolor='black', linewidth=3)
        ax.add_patch(clock_face)
        
        # ç»˜åˆ¶åˆ»åº¦
        for j in range(12):
            angle = j * np.pi / 6
            x1 = 0.8 * np.cos(angle)
            y1 = 0.8 * np.sin(angle)
            x2 = 0.85 * np.cos(angle)
            y2 = 0.85 * np.sin(angle)
            ax.plot([x1, x2], [y1, y2], 'black', linewidth=2)
            
            # æ·»åŠ æ•°å­—
            num_x = 0.65 * np.cos(angle - np.pi/2)
            num_y = 0.65 * np.sin(angle - np.pi/2)
            num = 12 if j == 0 else j
            ax.text(num_x, num_y, str(num), fontsize=10, ha='center', va='center', fontweight='bold')
        
        # è®¡ç®—æ—¶é’ˆè§’åº¦
        hour_angle = (time['hour'] % 12 + time['minute'] / 60) * np.pi / 6 - np.pi/2
        hour_x = 0.45 * np.cos(hour_angle)
        hour_y = 0.45 * np.sin(hour_angle)
        ax.plot([0, hour_x], [0, hour_y], 'red', linewidth=5)
        
        # è®¡ç®—åˆ†é’ˆè§’åº¦
        minute_angle = time['minute'] * np.pi / 30 - np.pi/2
        minute_x = 0.65 * np.cos(minute_angle)
        minute_y = 0.65 * np.sin(minute_angle)
        ax.plot([0, minute_x], [0, minute_y], 'blue', linewidth=3)
        
        # ç»˜åˆ¶ä¸­å¿ƒç‚¹
        center = plt.Circle((0, 0), 0.04, facecolor='black', edgecolor='black')
        ax.add_patch(center)
        
        # æ ‡é¢˜
        ax.text(0, -1.2, f"ç¬¬{i+1}å", fontsize=12, ha='center', fontweight='bold')
        ax.text(0, 1.2, time['label'], fontsize=12, ha='center', fontweight='bold')
        
        ax.set_xlim(-1.3, 1.3)
        ax.set_ylim(-1.3, 1.3)
        ax.set_aspect('equal')
        ax.axis('off')
    
    plt.suptitle('æŒ‰æ—¶é—´ä»æ—©åˆ°æ™šæ’åº', fontsize=16, fontweight='bold', y=1.05)
    plt.tight_layout()
    plt.show()

# è¿è¡Œæ—¶é—´æ’åº
sort_times()
```

## 7. æ—¶é—´é•¿çŸ­æ¯”è¾ƒ

å­¦ä¹ æ¯”è¾ƒæ—¶é—´çš„é•¿çŸ­ï¼

```python
def compare_time_duration():
    """æ¯”è¾ƒæ—¶é—´é•¿çŸ­"""
    print("â±ï¸ æ—¶é—´é•¿çŸ­æ¯”è¾ƒ")
    print("=" * 40)
    
    comparisons = [
        {'time1': '1å°æ—¶', 'time2': '30åˆ†é’Ÿ', 'longer': '1å°æ—¶', 'reason': '1å°æ—¶ = 60åˆ†é’Ÿï¼Œæ¯”30åˆ†é’Ÿé•¿'},
        {'time1': '1å°æ—¶', 'time2': '2å°æ—¶', 'longer': '2å°æ—¶', 'reason': '2å°æ—¶æ¯”1å°æ—¶é•¿'},
        {'time1': '30åˆ†é’Ÿ', 'time2': '1å°æ—¶', 'longer': '1å°æ—¶', 'reason': '1å°æ—¶ = 60åˆ†é’Ÿï¼Œæ¯”30åˆ†é’Ÿé•¿'},
        {'time1': '1å°æ—¶', 'time2': '1å°æ—¶', 'longer': 'ä¸€æ ·é•¿', 'reason': 'éƒ½æ˜¯1å°æ—¶ï¼Œæ—¶é—´ç›¸åŒ'},
        {'time1': '2å°æ—¶', 'time2': '1å°æ—¶', 'longer': '2å°æ—¶', 'reason': '2å°æ—¶æ¯”1å°æ—¶é•¿'}
    ]
    
    score = 0
    
    for i, comp in enumerate(comparisons):
        print(f"\né—®é¢˜ {i+1}:")
        print(f"{comp['time1']} å’Œ {comp['time2']}ï¼Œå“ªä¸ªæ—¶é—´é•¿ï¼Ÿ")
        
        answer = input("ä½ çš„ç­”æ¡ˆï¼š")
        
        if answer == comp['longer']:
            print("âœ… ç­”å¯¹äº†ï¼")
            print(f"åŸå› ï¼š{comp['reason']}")
            score += 1
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {comp['longer']}")
            print(f"åŸå› ï¼š{comp['reason']}")
    
    print(f"\nğŸ¯ ä½ çš„å¾—åˆ†ï¼š{score}/{len(comparisons)}")
    
    if score == len(comparisons):
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= len(comparisons) * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ—¶é—´æ¯”è¾ƒ
compare_time_duration()
```
