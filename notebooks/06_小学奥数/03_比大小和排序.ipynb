## 1. è®¤è¯†æ¯”è¾ƒç¬¦å·

è®©æˆ‘ä»¬è®¤è¯†ä¸‰ä¸ªé‡è¦çš„ç¬¦å·ï¼š
- **>** å¤§äºå·ï¼šå·¦è¾¹çš„æ•°æ¯”å³è¾¹çš„æ•°å¤§
- **<** å°äºå·ï¼šå·¦è¾¹çš„æ•°æ¯”å³è¾¹çš„æ•°å°
- **=** ç­‰äºå·ï¼šä¸¤è¾¹çš„æ•°ä¸€æ ·å¤§

```python
def visualize_comparison(a, b):
    """å¯è§†åŒ–æ•°å­—æ¯”è¾ƒ"""
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # ç»˜åˆ¶ç¬¬ä¸€ä¸ªæ•°çš„åœ†åœˆ
    for i in range(a):
        ax.add_patch(plt.Circle((i*1.2 + 0.5, 0.7), 0.4, 
                               facecolor='lightblue', edgecolor='blue', linewidth=2))
        ax.text(i*1.2 + 0.5, 0.7, str(i+1), fontsize=14, ha='center', va='center', fontweight='bold')
    
    # æ¯”è¾ƒç¬¦å·
    if a > b:
        symbol = '>'
        color = 'red'
        text = 'å¤§äº'
    elif a < b:
        symbol = '<'
        color = 'green'
        text = 'å°äº'
    else:
        symbol = '='
        color = 'orange'
        text = 'ç­‰äº'
    
    ax.text(a*1.2 + 1.0, 0.7, symbol, fontsize=60, ha='center', va='center', 
            fontweight='bold', color=color)
    
    # ç»˜åˆ¶ç¬¬äºŒä¸ªæ•°çš„åœ†åœˆ
    for i in range(b):
        ax.add_patch(plt.Circle((a*1.2 + 2.0 + i*1.2, 0.7), 0.4, 
                               facecolor='lightgreen', edgecolor='green', linewidth=2))
        ax.text(a*1.2 + 2.0 + i*1.2, 0.7, str(i+1), fontsize=14, ha='center', va='center', fontweight='bold')
    
    # æ˜¾ç¤ºæ•°å­—
    ax.text(a*1.2 + 1.0, 0.3, f'{a} {symbol} {b}', fontsize=24, ha='center', 
            va='center', fontweight='bold', color=color)
    ax.text(a*1.2 + 1.0, 0.15, f'{a} {text} {b}', fontsize=16, ha='center', va='center')
    
    ax.set_xlim(0, (a+b)*1.2 + 3.0)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# å±•ç¤ºæ¯”è¾ƒ
print("ğŸ“ æ•°å­—æ¯”è¾ƒæ¼”ç¤ºï¼š")
visualize_comparison(5, 3)
visualize_comparison(2, 7)
visualize_comparison(4, 4)
```

## 2. æ¯”å¤§å°æ¸¸æˆ

æ¥ç©æ¯”å¤§å°æ¸¸æˆå§ï¼

```python
import random

def compare_numbers_game():
    """æ¯”å¤§å°æ¸¸æˆ"""
    print("ğŸ® æ¯”å¤§å°æ¸¸æˆ")
    print("=" * 40)
    print("è§„åˆ™ï¼šåˆ¤æ–­ä¸¤ä¸ªæ•°çš„å¤§å°å…³ç³»")
    print("è¾“å…¥ '>' è¡¨ç¤ºå¤§äºï¼Œ'<' è¡¨ç¤ºå°äºï¼Œ'=' è¡¨ç¤ºç­‰äº")
    
    score = 0
    total_questions = 5
    
    for i in range(total_questions):
        print(f"\né—®é¢˜ {i+1}/{total_questions}")
        
        a = random.randint(1, 10)
        b = random.randint(1, 10)
        
        print(f"{a} ___ {b}")
        
        answer = input("å¡«å…¥ >ã€< æˆ– =ï¼š")
        
        correct_answer = ''
        if a > b:
            correct_answer = '>'
        elif a < b:
            correct_answer = '<'
        else:
            correct_answer = '='
        
        if answer == correct_answer:
            print("âœ… ç­”å¯¹äº†ï¼")
            score += 1
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {correct_answer}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{total_questions}")
    
    if score == total_questions:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= total_questions * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
compare_numbers_game()
```

## 3. æ•°å­—æ’åº

å­¦ä¹ å¦‚ä½•æŠŠæ•°å­—æŒ‰ä»å°åˆ°å¤§æˆ–ä»å¤§åˆ°å°æ’åˆ—ï¼

```python
def visualize_sorting(numbers, ascending=True):
    """å¯è§†åŒ–æ•°å­—æ’åº"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # åŸå§‹é¡ºåº
    for i, num in enumerate(numbers):
        ax1.add_patch(plt.Rectangle((i*0.8 + 0.2, 0.2), 0.6, num*0.08, 
                                     facecolor='lightblue', edgecolor='blue', linewidth=2))
        ax1.text(i*0.8 + 0.5, num*0.08 + 0.3, str(num), fontsize=14, 
                ha='center', va='bottom', fontweight='bold')
    
    ax1.set_xlim(0, len(numbers)*0.8 + 0.4)
    ax1.set_ylim(0, 1.2)
    ax1.set_aspect('equal')
    ax1.axis('off')
    ax1.set_title('åŸå§‹é¡ºåº', fontsize=16, fontweight='bold')
    
    # æ’åºåçš„é¡ºåº
    sorted_numbers = sorted(numbers, reverse=not ascending)
    order_text = 'ä»å°åˆ°å¤§' if ascending else 'ä»å¤§åˆ°å°'
    
    for i, num in enumerate(sorted_numbers):
        if ascending:
            color = plt.cm.RdYlGn(i / len(sorted_numbers))
        else:
            color = plt.cm.RdYlGn_r(i / len(sorted_numbers))
        
        ax2.add_patch(plt.Rectangle((i*0.8 + 0.2, 0.2), 0.6, num*0.08, 
                                     facecolor=color, edgecolor='black', linewidth=2))
        ax2.text(i*0.8 + 0.5, num*0.08 + 0.3, str(num), fontsize=14, 
                ha='center', va='bottom', fontweight='bold')
    
    ax2.set_xlim(0, len(sorted_numbers)*0.8 + 0.4)
    ax2.set_ylim(0, 1.2)
    ax2.set_aspect('equal')
    ax2.axis('off')
    ax2.set_title(f'æ’åºåï¼ˆ{order_text}ï¼‰', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    return sorted_numbers

# æ’åºæ¼”ç¤º
print("ğŸ“Š æ•°å­—æ’åºæ¼”ç¤ºï¼š")
numbers = [3, 7, 2, 9, 5, 1, 8, 4, 6]
print(f"åŸå§‹æ•°å­—ï¼š{numbers}")

sorted_asc = visualize_sorting(numbers, ascending=True)
print(f"ä»å°åˆ°å¤§ï¼š{sorted_asc}")

sorted_desc = visualize_sorting(numbers, ascending=False)
print(f"ä»å¤§åˆ°å°ï¼š{sorted_desc}")
```

## 4. æ‰¾æœ€å¤§å’Œæœ€å°

åœ¨ä¸€ç»„æ•°å­—ä¸­æ‰¾å‡ºæœ€å¤§å’Œæœ€å°çš„æ•°ï¼

```python
def find_max_min(numbers):
    """æ‰¾å‡ºæœ€å¤§å’Œæœ€å°å€¼"""
    fig, ax = plt.subplots(figsize=(14, 6))
    
    max_val = max(numbers)
    min_val = min(numbers)
    
    for i, num in enumerate(numbers):
        if num == max_val:
            color = 'red'
            label = 'æœ€å¤§'
            edge_width = 4
        elif num == min_val:
            color = 'green'
            label = 'æœ€å°'
            edge_width = 4
        else:
            color = 'lightblue'
            label = ''
            edge_width = 2
        
        ax.add_patch(plt.Rectangle((i*0.8 + 0.2, 0.2), 0.6, num*0.08, 
                                     facecolor=color, edgecolor='black', linewidth=edge_width))
        ax.text(i*0.8 + 0.5, num*0.08 + 0.3, str(num), fontsize=14, 
                ha='center', va='bottom', fontweight='bold')
        
        if label:
            ax.text(i*0.8 + 0.5, 0.1, label, fontsize=12, ha='center', va='center', 
                    fontweight='bold', color=color)
    
    ax.set_xlim(0, len(numbers)*0.8 + 0.4)
    ax.set_ylim(0, 1.2)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(f'æœ€å¤§å€¼ï¼š{max_val}ï¼Œæœ€å°å€¼ï¼š{min_val}', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    return max_val, min_val

# æ‰¾æœ€å¤§æœ€å°
print("ğŸ” æ‰¾æœ€å¤§å’Œæœ€å°ï¼š")
numbers = [4, 9, 2, 7, 5, 1, 8, 3, 6]
print(f"æ•°å­—ï¼š{numbers}")
max_val, min_val = find_max_min(numbers)
print(f"æœ€å¤§å€¼æ˜¯ {max_val}ï¼Œæœ€å°å€¼æ˜¯ {min_val}")
```

## 6. æ’åºæ¸¸æˆ

æ¥ç©æ’åºæ¸¸æˆå§ï¼

```python
def sorting_game():
    """æ’åºæ¸¸æˆ"""
    print("ğŸ® æ’åºæ¸¸æˆ")
    print("=" * 40)
    
    score = 0
    total_questions = 3
    
    for i in range(total_questions):
        print(f"\né—®é¢˜ {i+1}/{total_questions}")
        
        # ç”Ÿæˆéšæœºæ•°å­—
        num_count = random.randint(3, 5)
        numbers = [random.randint(1, 10) for _ in range(num_count)]
        
        print(f"æ•°å­—ï¼š{numbers}")
        
        # éšæœºé€‰æ‹©æ’åºæ–¹å¼
        if random.choice([True, False]):
            order = 'ä»å°åˆ°å¤§'
            correct_answer = sorted(numbers)
        else:
            order = 'ä»å¤§åˆ°å°'
            correct_answer = sorted(numbers, reverse=True)
        
        print(f"è¯·æŒ‰{order}çš„é¡ºåºæ’åˆ—è¿™äº›æ•°å­—")
        
        answer = input("ä½ çš„ç­”æ¡ˆï¼ˆç”¨ç©ºæ ¼åˆ†éš”ï¼‰ï¼š")
        
        try:
            user_answer = [int(x) for x in answer.split()]
            if user_answer == correct_answer:
                print("âœ… ç­”å¯¹äº†ï¼")
                score += 1
            else:
                print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {correct_answer}")
        except:
            print(f"âŒ æ ¼å¼ä¸å¯¹ï¼Œè¯·ç”¨ç©ºæ ¼åˆ†éš”æ•°å­—ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {correct_answer}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{total_questions}")
    
    if score == total_questions:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= total_questions * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
sorting_game()
```

## 7. æ•°å­—æ¯”è¾ƒè¡¨

çœ‹çœ‹1åˆ°10çš„æ‰€æœ‰æ•°å­—æ¯”è¾ƒï¼

```python
def create_comparison_table():
    """åˆ›å»ºæ•°å­—æ¯”è¾ƒè¡¨"""
    fig, ax = plt.subplots(figsize=(14, 14))
    
    size = 11
    
    # ç»˜åˆ¶æ¯”è¾ƒè¡¨
    for i in range(size):
        for j in range(size):
            if i == 0 and j == 0:
                text = 'â†˜'
                color = 'lightgray'
            elif i == 0:
                text = str(j)
                color = 'lightyellow'
            elif j == 0:
                text = str(i)
                color = 'lightyellow'
            else:
                if i > j:
                    text = '>'
                    color = 'lightcoral'
                elif i < j:
                    text = '<'
                    color = 'lightgreen'
                else:
                    text = '='
                    color = 'lightblue'
            
            # ç»˜åˆ¶å•å…ƒæ ¼
            rect = plt.Rectangle((j, size-1-i), 1, 1, 
                                 facecolor=color, edgecolor='black', linewidth=1)
            ax.add_patch(rect)
            
            # æ·»åŠ ç¬¦å·
            ax.text(j + 0.5, size - 0.5 - i, text, 
                   fontsize=20, ha='center', va='center', fontweight='bold')
    
    ax.set_xlim(0, size)
    ax.set_ylim(0, size)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('1-10æ•°å­—æ¯”è¾ƒè¡¨ï¼ˆè¡Œ > åˆ—ï¼‰', fontsize=18, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

# æ˜¾ç¤ºæ¯”è¾ƒè¡¨
create_comparison_table()
```

## 8. æ‰¾ç¬¬å‡ å

å­¦ä¹ æ‰¾å‡ºæ•°å­—ä¸­çš„ç¬¬å‡ åï¼

```python
def find_rank(numbers, rank):
    """æ‰¾å‡ºç¬¬å‡ åçš„æ•°å­—"""
    sorted_numbers = sorted(numbers)
    if 1 <= rank <= len(sorted_numbers):
        return sorted_numbers[rank - 1]
    else:
        return None

def visualize_rank(numbers, rank):
    """å¯è§†åŒ–æ’å"""
    sorted_numbers = sorted(numbers)
    target = sorted_numbers[rank - 1]
    
    fig, ax = plt.subplots(figsize=(14, 6))
    
    for i, num in enumerate(sorted_numbers):
        if i + 1 == rank:
            color = 'red'
            edge_width = 4
            label = f'ç¬¬{rank}å'
        else:
            color = 'lightblue'
            edge_width = 2
            label = ''
        
        ax.add_patch(plt.Rectangle((i*0.8 + 0.2, 0.2), 0.6, num*0.08, 
                                     facecolor=color, edgecolor='black', linewidth=edge_width))
        ax.text(i*0.8 + 0.5, num*0.08 + 0.3, str(num), fontsize=14, 
                ha='center', va='bottom', fontweight='bold')
        ax.text(i*0.8 + 0.5, 0.1, f'ç¬¬{i+1}å', fontsize=10, ha='center', va='center')
        
        if label:
            ax.text(i*0.8 + 0.5, num*0.08 + 0.5, label, fontsize=12, 
                    ha='center', va='bottom', fontweight='bold', color='red')
    
    ax.set_xlim(0, len(sorted_numbers)*0.8 + 0.4)
    ax.set_ylim(0, 1.2)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(f'ç¬¬{rank}åæ˜¯ï¼š{target}', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    return target

# æ‰¾ç¬¬å‡ å
print("ğŸ† æ‰¾ç¬¬å‡ åï¼š")
numbers = [3, 8, 1, 6, 9, 2, 7, 4, 5]
print(f"æ•°å­—ï¼š{numbers}")

rank_1 = visualize_rank(numbers, 1)
print(f"ç¬¬1åï¼ˆæœ€å°ï¼‰æ˜¯ï¼š{rank_1}")

rank_3 = visualize_rank(numbers, 3)
print(f"ç¬¬3åæ˜¯ï¼š{rank_3}")

rank_last = visualize_rank(numbers, len(numbers))
print(f"ç¬¬{len(numbers)}åï¼ˆæœ€å¤§ï¼‰æ˜¯ï¼š{rank_last}")
```
