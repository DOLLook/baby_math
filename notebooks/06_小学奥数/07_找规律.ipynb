## 1. 什么是规律

规律就是按照一定顺序重复出现的模式！

```python
import matplotlib.pyplot as plt
import numpy as np

def explain_pattern():
    """解释什么是规律"""
    print("🔍 什么是规律？")
    print("=" * 40)
    print("规律就是按照一定顺序重复出现的模式！")
    print("\n例子1：")
    print("1, 2, 1, 2, 1, 2, ...")
    print("规律是：1, 2 重复出现")
    
    print("\n例子2：")
    print("红, 蓝, 红, 蓝, 红, 蓝, ...")
    print("规律是：红, 蓝重复出现")
    
    # 可视化例子
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # 数字规律
    ax = axes[0]
    numbers = [1, 2, 1, 2, 1, 2]
    for i, num in enumerate(numbers):
        color = 'red' if num == 1 else 'blue'
        ax.text(i, 0, str(num), fontsize=20, ha='center', va='center', 
                bbox=dict(facecolor=color, alpha=0.3, edgecolor=color, linewidth=2))
    ax.text(3, 1, '数字规律', fontsize=14, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1, 1.5)
    ax.axis('off')
    
    # 颜色规律
    ax = axes[1]
    colors = ['red', 'blue', 'red', 'blue', 'red', 'blue']
    for i, color in enumerate(colors):
        circle = plt.Circle((i, 0), 0.3, facecolor=color, edgecolor=color, linewidth=2)
        ax.add_patch(circle)
    ax.text(2.5, 1, '颜色规律', fontsize=14, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.tight_layout()
    plt.show()

# 运行解释
explain_pattern()
```

## 2. 找数字规律

让我们来找找数字的规律！

```python
def find_number_pattern():
    """找数字规律"""
    print("🔢 找数字规律")
    print("=" * 40)
    
    patterns = [
        {
            'sequence': [1, 2, 3, 4, 5, '?'],
            'pattern': '每次加1',
            'next': 6
        },
        {
            'sequence': [2, 4, 6, 8, 10, '?'],
            'pattern': '每次加2',
            'next': 12
        },
        {
            'sequence': [1, 3, 5, 7, 9, '?'],
            'pattern': '每次加2（奇数）',
            'next': 11
        },
        {
            'sequence': [10, 9, 8, 7, 6, '?'],
            'pattern': '每次减1',
            'next': 5
        },
        {
            'sequence': [1, 2, 1, 2, 1, '?'],
            'pattern': '1, 2 重复',
            'next': 2
        }
    ]
    
    score = 0
    
    for i, pattern in enumerate(patterns):
        print(f"\n问题 {i+1}:")
        print(f"数列：{', '.join(map(str, pattern['sequence']))}")
        print(f"规律是什么？")
        
        answer = input("你的答案：")
        
        if answer == pattern['pattern']:
            print("✅ 答对了！")
            print(f"下一个数是：{pattern['next']}")
            score += 1
        else:
            print(f"❌ 不对哦，规律是：{pattern['pattern']}")
            print(f"下一个数是：{pattern['next']}")
    
    print(f"\n🎯 你的得分：{score}/{len(patterns)}")
    
    if score == len(patterns):
        print("🏆 太厉害了！全部正确！")
    elif score >= len(patterns) * 0.8:
        print("🌟 做得很好！继续保持！")
    else:
        print("💪 继续加油，多练习就会进步！")

# 运行找数字规律
find_number_pattern()
```

## 3. 找图形规律

让我们来找找图形的规律！

```python
def find_shape_pattern():
    """找图形规律"""
    print("🔷 找图形规律")
    print("=" * 40)
    
    # 创建图形规律
    fig, axes = plt.subplots(3, 2, figsize=(12, 15))
    
    # 规律1：圆形、正方形重复
    ax = axes[0, 0]
    for i in range(5):
        if i % 2 == 0:
            circle = plt.Circle((i, 0), 0.3, facecolor='lightblue', 
                               edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
        else:
            square = plt.Rectangle((i-0.25, -0.25), 0.5, 0.5, facecolor='lightgreen', 
                                  edgecolor='green', linewidth=2)
            ax.add_patch(square)
    ax.text(2, 1, '规律：圆形、正方形重复', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 规律2：三角形、长方形重复
    ax = axes[0, 1]
    for i in range(5):
        if i % 2 == 0:
            triangle = plt.Polygon([(i-0.3, -0.25), (i+0.3, -0.25), (i, 0.25)], 
                                  facecolor='lightyellow', edgecolor='orange', linewidth=2)
            ax.add_patch(triangle)
        else:
            rectangle = plt.Rectangle((i-0.4, -0.25), 0.8, 0.5, facecolor='lightpink', 
                                     edgecolor='red', linewidth=2)
            ax.add_patch(rectangle)
    ax.text(2, 1, '规律：三角形、长方形重复', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 规律3：颜色变化
    ax = axes[1, 0]
    colors = ['red', 'blue', 'green', 'red', 'blue']
    for i, color in enumerate(colors):
        circle = plt.Circle((i, 0), 0.3, facecolor=color, edgecolor=color, linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, '规律：红、蓝、绿重复', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 规律4：大小变化
    ax = axes[1, 1]
    for i in range(5):
        size = 0.2 + i * 0.1
        circle = plt.Circle((i, 0), size, facecolor='lightblue', 
                           edgecolor='blue', linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, '规律：越来越大', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 规律5：位置变化
    ax = axes[2, 0]
    for i in range(5):
        y = 0.3 if i % 2 == 0 else -0.3
        circle = plt.Circle((i, y), 0.3, facecolor='lightgreen', 
                           edgecolor='green', linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, '规律：上下交替', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 规律6：数量变化
    ax = axes[2, 1]
    for i in range(5):
        for j in range(i + 1):
            circle = plt.Circle((i, j * 0.3 - 0.6), 0.1, facecolor='lightpink', 
                               edgecolor='red', linewidth=2)
            ax.add_patch(circle)
    ax.text(2, 1, '规律：数量递增', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '下一个是？', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.suptitle('找图形规律', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # 问答
    print("\n🤔 思考题：")
    print("1. 圆形、正方形重复，下一个是什么？")
    print("   答案：圆形")
    print("2. 三角形、长方形重复，下一个是什么？")
    print("   答案：三角形")
    print("3. 红、蓝、绿重复，下一个是什么颜色？")
    print("   答案：绿色")
    print("4. 越来越大，下一个是什么？")
    print("   答案：更大的圆")
    print("5. 上下交替，下一个在哪里？")
    print("   答案：上面")
    print("6. 数量递增，下一个有几个？")
    print("   答案：5个")

# 运行找图形规律
find_shape_pattern()
```

## 4. 数学故事：小松鼠找规律

通过故事学习找规律！

```python
def squirrel_pattern_story():
    """小松鼠找规律的故事"""
    print("📖 小松鼠找规律")
    print("=" * 40)
    
    print("\n从前，有一只聪明的小松鼠，名叫小松。")
    print("小松最喜欢找规律了！")
    
    print("\n有一天，小松在森林里发现了一串松果：")
    print("1, 2, 3, 4, 5, ...")
    print("小松想：'这个规律是每次加1！'")
    print("'下一个应该是6！'")
    
    print("\n接着，小松又发现了一串彩色的树叶：")
    print("红, 黄, 红, 黄, 红, ...")
    print("小松想：'这个规律是红、黄重复！'")
    print("'下一个应该是黄色！'")
    
    print("\n小松继续往前走，发现了一串脚印：")
    print("大, 小, 大, 小, 大, ...")
    print("小松想：'这个规律是大、小重复！'")
    print("'下一个应该是小脚印！'")
    
    print("\n小松高兴地跳了起来：")
    print("'我找到规律啦！找规律真好玩！'")
    
    # 可视化故事
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # 松果规律
    ax = axes[0]
    for i in range(5):
        circle = plt.Circle((i, 0), 0.2 + i * 0.05, facecolor='brown', 
                           edgecolor='brown', linewidth=2)
        ax.add_patch(circle)
        ax.text(i, 0, str(i+1), fontsize=12, ha='center', va='center', color='white', fontweight='bold')
    ax.text(2, 1, '松果规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2, -1, '每次加1', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 树叶规律
    ax = axes[1]
    colors = ['red', 'yellow', 'red', 'yellow', 'red']
    for i, color in enumerate(colors):
        ellipse = plt.Ellipse((i, 0), 0.4, 0.2, facecolor=color, 
                              edgecolor=color, linewidth=2)
        ax.add_patch(ellipse)
    ax.text(2, 1, '树叶规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2, -1, '红、黄重复', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 脚印规律
    ax = axes[2]
    for i in range(5):
        size = 0.3 if i % 2 == 0 else 0.15
        circle = plt.Circle((i, 0), size, facecolor='gray', 
                           edgecolor='black', linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, '脚印规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2, -1, '大、小重复', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.suptitle('小松鼠找规律', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # 思考题
    print("\n🤔 思考题：")
    print("1. 松果规律的下一个是什么？")
    print("   答案：6")
    print("2. 树叶规律的下一个是什么颜色？")
    print("   答案：黄色")
    print("3. 脚印规律的下一个是什么？")
    print("   答案：小脚印")

# 运行故事
squirrel_pattern_story()
```

## 5. 找规律游戏

来玩找规律的游戏吧！

```python
import random

def pattern_game():
    """找规律游戏"""
    print("🎮 找规律游戏")
    print("=" * 40)
    print("规则：看规律，说出下一个是什么")
    
    score = 0
    total_questions = 5
    
    # 创建各种规律
    patterns = [
        {
            'type': 'number',
            'sequence': [1, 2, 3, 4, 5],
            'pattern': '每次加1',
            'next': 6
        },
        {
            'type': 'number',
            'sequence': [2, 4, 6, 8, 10],
            'pattern': '每次加2',
            'next': 12
        },
        {
            'type': 'number',
            'sequence': [10, 9, 8, 7, 6],
            'pattern': '每次减1',
            'next': 5
        },
        {
            'type': 'number',
            'sequence': [1, 2, 1, 2, 1],
            'pattern': '1, 2重复',
            'next': 2
        },
        {
            'type': 'number',
            'sequence': [1, 1, 2, 2, 3],
            'pattern': '每个数重复2次',
            'next': 3
        }
    ]
    
    # 随机选择5个问题
    selected_patterns = random.sample(patterns, min(total_questions, len(patterns)))
    
    for i, pattern in enumerate(selected_patterns):
        print(f"\n问题 {i+1}/{total_questions}")
        
        # 显示规律
        print(f"数列：{', '.join(map(str, pattern['sequence']))}")
        
        # 询问规律
        pattern_answer = input("规律是什么？")
        
        if pattern_answer == pattern['pattern']:
            print("✅ 规律答对了！")
            
            # 询问下一个
            next_answer = input(f"下一个数是什么？")
            
            try:
                if int(next_answer) == pattern['next']:
                    print("✅ 全部正确！")
                    score += 1
                else:
                    print(f"❌ 不对哦，下一个数是 {pattern['next']}")
            except:
                print(f"❌ 不对哦，下一个数是 {pattern['next']}")
        else:
            print(f"❌ 不对哦，规律是：{pattern['pattern']}")
            print(f"下一个数是：{pattern['next']}")
    
    print(f"\n🎯 游戏结束！")
    print(f"你的得分：{score}/{total_questions}")
    
    if score == total_questions:
        print("🏆 太厉害了！全部正确！")
    elif score >= total_questions * 0.8:
        print("🌟 做得很好！继续保持！")
    else:
        print("💪 继续加油，多练习就会进步！")

# 运行游戏
pattern_game()
```

## 6. 创造规律

让我们来创造自己的规律吧！

```python
def create_pattern():
    """创造规律"""
    print("🎨 创造规律")
    print("=" * 40)
    print("让我们来创造自己的规律吧！")
    
    # 示例：创造一个数字规律
    print("\n示例1：数字规律")
    sequence = [1, 3, 5, 7, 9, 11]
    print(f"规律：{', '.join(map(str, sequence))}")
    print("说明：每次加2（奇数）")
    
    # 示例：创造一个颜色规律
    print("\n示例2：颜色规律")
    colors = ['红', '蓝', '绿', '红', '蓝', '绿']
    print(f"规律：{', '.join(colors)}")
    print("说明：红、蓝、绿重复")
    
    # 示例：创造一个图形规律
    print("\n示例3：图形规律")
    shapes = ['圆形', '正方形', '三角形', '圆形', '正方形', '三角形']
    print(f"规律：{', '.join(shapes)}")
    print("说明：圆形、正方形、三角形重复")
    
    # 可视化示例
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # 数字规律
    ax = axes[0]
    for i, num in enumerate(sequence):
        color = 'red' if num % 2 == 1 else 'blue'
        ax.text(i, 0, str(num), fontsize=16, ha='center', va='center', 
                bbox=dict(facecolor=color, alpha=0.3, edgecolor=color, linewidth=2))
    ax.text(2.5, 1, '数字规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2.5, -1, '每次加2', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1.5, 1.5)
    ax.axis('off')
    
    # 颜色规律
    ax = axes[1]
    color_map = {'红': 'red', '蓝': 'blue', '绿': 'green'}
    for i, color in enumerate(colors):
        circle = plt.Circle((i, 0), 0.3, facecolor=color_map[color], 
                           edgecolor=color_map[color], linewidth=2)
        ax.add_patch(circle)
    ax.text(2.5, 1, '颜色规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2.5, -1, '红、蓝、绿重复', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # 图形规律
    ax = axes[2]
    for i, shape in enumerate(shapes):
        if shape == '圆形':
            circle = plt.Circle((i, 0), 0.25, facecolor='lightblue', 
                               edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
        elif shape == '正方形':
            square = plt.Rectangle((i-0.2, -0.2), 0.4, 0.4, facecolor='lightgreen', 
                                  edgecolor='green', linewidth=2)
            ax.add_patch(square)
        elif shape == '三角形':
            triangle = plt.Polygon([(i-0.2, -0.2), (i+0.2, -0.2), (i, 0.2)], 
                                  facecolor='lightyellow', edgecolor='orange', linewidth=2)
            ax.add_patch(triangle)
    ax.text(2.5, 1, '图形规律', fontsize=14, ha='center', fontweight='bold')
    ax.text(2.5, -1, '圆形、正方形、三角形重复', fontsize=10, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.suptitle('创造规律示例', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # 让用户创造自己的规律
    print("\n🎯 现在轮到你了！")
    print("请创造你自己的规律：")
    
    pattern_type = input("你想创造什么类型的规律？（数字/颜色/图形）")
    
    if pattern_type == '数字':
        print("请输入6个数字，用空格分隔：")
        numbers = list(map(int, input().split()))
        print(f"你的规律：{', '.join(map(str, numbers))}")
        pattern_desc = input("请描述你的规律：")
        print(f"规律说明：{pattern_desc}")
        print("🎉 太棒了！你创造了自己的规律！")
    elif pattern_type == '颜色':
        print("请输入6种颜色，用空格分隔：")
        colors = input().split()
        print(f"你的规律：{', '.join(colors)}")
        pattern_desc = input("请描述你的规律：")
        print(f"规律说明：{pattern_desc}")
        print("🎉 太棒了！你创造了自己的规律！")
    elif pattern_type == '图形':
        print("请输入6种图形（圆形/正方形/三角形），用空格分隔：")
        shapes = input().split()
        print(f"你的规律：{', '.join(shapes)}")
        pattern_desc = input("请描述你的规律：")
        print(f"规律说明：{pattern_desc}")
        print("🎉 太棒了！你创造了自己的规律！")
    else:
        print("请选择：数字、颜色或图形")

# 运行创造规律
create_pattern()
```
