```python
def find_number_pattern():
    """æ‰¾æ•°å­—è§„å¾‹"""
    print("ğŸ”¢ æ‰¾æ•°å­—è§„å¾‹")
    print("=" * 40)
    
    patterns = [
        {
            'sequence': [1, 2, 3, 4, 5, '?'],
            'pattern': 'æ¯æ¬¡åŠ 1',
            'next': 6
        },
        {
            'sequence': [2, 4, 6, 8, 10, '?'],
            'pattern': 'æ¯æ¬¡åŠ 2',
            'next': 12
        },
        {
            'sequence': [1, 3, 5, 7, 9, '?'],
            'pattern': 'æ¯æ¬¡åŠ 2ï¼ˆå¥‡æ•°ï¼‰',
            'next': 11
        },
        {
            'sequence': [10, 9, 8, 7, 6, '?'],
            'pattern': 'æ¯æ¬¡å‡1',
            'next': 5
        },
        {
            'sequence': [1, 2, 1, 2, 1, '?'],
            'pattern': '1, 2 é‡å¤',
            'next': 2
        }
    ]
    
    score = 0
    
    for i, pattern in enumerate(patterns):
        print(f"\né—®é¢˜ {i+1}:")
        print(f"æ•°åˆ—ï¼š{', '.join(map(str, pattern['sequence']))}")
        print(f"è§„å¾‹æ˜¯ä»€ä¹ˆï¼Ÿ")
        
        answer = input("ä½ çš„ç­”æ¡ˆï¼š")
        
        if answer == pattern['pattern']:
            print("âœ… ç­”å¯¹äº†ï¼")
            print(f"ä¸‹ä¸€ä¸ªæ•°æ˜¯ï¼š{pattern['next']}")
            score += 1
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œè§„å¾‹æ˜¯ï¼š{pattern['pattern']}")
            print(f"ä¸‹ä¸€ä¸ªæ•°æ˜¯ï¼š{pattern['next']}")
    
    print(f"\nğŸ¯ ä½ çš„å¾—åˆ†ï¼š{score}/{len(patterns)}")
    
    if score == len(patterns):
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= len(patterns) * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ‰¾æ•°å­—è§„å¾‹
find_number_pattern()
```

## 5. æ‰¾è§„å¾‹æ¸¸æˆ

æ¥ç©æ‰¾è§„å¾‹çš„æ¸¸æˆå§ï¼

```python
import random

def pattern_game():
    """æ‰¾è§„å¾‹æ¸¸æˆ"""
    print("ğŸ® æ‰¾è§„å¾‹æ¸¸æˆ")
    print("=" * 40)
    print("è§„åˆ™ï¼šçœ‹è§„å¾‹ï¼Œè¯´å‡ºä¸‹ä¸€ä¸ªæ˜¯ä»€ä¹ˆ")
    
    score = 0
    total_questions = 5
    
    # åˆ›å»ºå„ç§è§„å¾‹
    patterns = [
        {
            'type': 'number',
            'sequence': [1, 2, 3, 4, 5],
            'pattern': 'æ¯æ¬¡åŠ 1',
            'next': 6
        },
        {
            'type': 'number',
            'sequence': [2, 4, 6, 8, 10],
            'pattern': 'æ¯æ¬¡åŠ 2',
            'next': 12
        },
        {
            'type': 'number',
            'sequence': [10, 9, 8, 7, 6],
            'pattern': 'æ¯æ¬¡å‡1',
            'next': 5
        },
        {
            'type': 'number',
            'sequence': [1, 2, 1, 2, 1],
            'pattern': '1, 2é‡å¤',
            'next': 2
        },
        {
            'type': 'number',
            'sequence': [1, 1, 2, 2, 3],
            'pattern': 'æ¯ä¸ªæ•°é‡å¤2æ¬¡',
            'next': 3
        }
    ]
    
    # éšæœºé€‰æ‹©5ä¸ªé—®é¢˜
    selected_patterns = random.sample(patterns, min(total_questions, len(patterns)))
    
    for i, pattern in enumerate(selected_patterns):
        print(f"\né—®é¢˜ {i+1}/{total_questions}")
        
        # æ˜¾ç¤ºè§„å¾‹
        print(f"æ•°åˆ—ï¼š{', '.join(map(str, pattern['sequence']))}")
        
        # è¯¢é—®è§„å¾‹
        pattern_answer = input("è§„å¾‹æ˜¯ä»€ä¹ˆï¼Ÿ")
        
        if pattern_answer == pattern['pattern']:
            print("âœ… è§„å¾‹ç­”å¯¹äº†ï¼")
            
            # è¯¢é—®ä¸‹ä¸€ä¸ª
            next_answer = input(f"ä¸‹ä¸€ä¸ªæ•°æ˜¯ä»€ä¹ˆï¼Ÿ")
            
            try:
                if int(next_answer) == pattern['next']:
                    print("âœ… å…¨éƒ¨æ­£ç¡®ï¼")
                    score += 1
                else:
                    print(f"âŒ ä¸å¯¹å“¦ï¼Œä¸‹ä¸€ä¸ªæ•°æ˜¯ {pattern['next']}")
            except:
                print(f"âŒ ä¸å¯¹å“¦ï¼Œä¸‹ä¸€ä¸ªæ•°æ˜¯ {pattern['next']}")
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œè§„å¾‹æ˜¯ï¼š{pattern['pattern']}")
            print(f"ä¸‹ä¸€ä¸ªæ•°æ˜¯ï¼š{pattern['next']}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{total_questions}")
    
    if score == total_questions:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= total_questions * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šç»ƒä¹ å°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
pattern_game()
```
