## 1. ‰ªÄ‰πàÊòØËßÑÂæã

```python
import matplotlib.pyplot as plt
import numpy as np

def explain_pattern():
    """Ëß£Èáä‰ªÄ‰πàÊòØËßÑÂæã"""
    print("üîç ‰ªÄ‰πàÊòØËßÑÂæãÔºü")
    print("=" * 40)
    print("ËßÑÂæãÂ∞±ÊòØÊåâÁÖß‰∏ÄÂÆöÈ°∫Â∫èÈáçÂ§çÂá∫Áé∞ÁöÑÊ®°ÂºèÔºÅ")
    print("\n‰æãÂ≠ê1Ôºö")
    print("1, 2, 1, 2, 1, 2, ...")
    print("ËßÑÂæãÊòØÔºö1, 2 ÈáçÂ§çÂá∫Áé∞")
    
    print("\n‰æãÂ≠ê2Ôºö")
    print("Á∫¢, Ëìù, Á∫¢, Ëìù, Á∫¢, Ëìù, ...")
    print("ËßÑÂæãÊòØÔºöÁ∫¢, ËìùÈáçÂ§çÂá∫Áé∞")
    
    # ÂèØËßÜÂåñ‰æãÂ≠ê
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Êï∞Â≠óËßÑÂæã
    ax = axes[0]
    numbers = [1, 2, 1, 2, 1, 2]
    for i, num in enumerate(numbers):
        color = 'red' if num == 1 else 'blue'
        ax.text(i, 0, str(num), fontsize=20, ha='center', va='center', 
                bbox=dict(facecolor=color, alpha=0.3, edgecolor=color, linewidth=2))
    ax.text(3, 1, 'Êï∞Â≠óËßÑÂæã', fontsize=14, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1, 1.5)
    ax.axis('off')
    
    # È¢úËâ≤ËßÑÂæã
    ax = axes[1]
    colors = ['red', 'blue', 'red', 'blue', 'red', 'blue']
    for i, color in enumerate(colors):
        circle = plt.Circle((i, 0), 0.3, facecolor=color, edgecolor=color, linewidth=2)
        ax.add_patch(circle)
    ax.text(2.5, 1, 'È¢úËâ≤ËßÑÂæã', fontsize=14, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 5.5)
    ax.set_ylim(-1, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.tight_layout()
    plt.show()

# ËøêË°åËß£Èáä
explain_pattern()
```

## 2. ÊâæÊï∞Â≠óËßÑÂæã

ËÆ©Êàë‰ª¨Êù•ÊâæÊâæÊï∞Â≠óÁöÑËßÑÂæãÔºÅ

```python
def find_number_pattern():
    """ÊâæÊï∞Â≠óËßÑÂæã"""
    print("üî¢ ÊâæÊï∞Â≠óËßÑÂæã")
    print("=" * 40)
    
    patterns = [
        {
            'sequence': [1, 2, 3, 4, 5, '?'],
            'pattern': 'ÊØèÊ¨°Âä†1',
            'next': 6
        },
        {
            'sequence': [2, 4, 6, 8, 10, '?'],
            'pattern': 'ÊØèÊ¨°Âä†2',
            'next': 12
        },
        {
            'sequence': [1, 3, 5, 7, 9, '?'],
            'pattern': 'ÊØèÊ¨°Âä†2ÔºàÂ•áÊï∞Ôºâ',
            'next': 11
        },
        {
            'sequence': [10, 9, 8, 7, 6, '?'],
            'pattern': 'ÊØèÊ¨°Âáè1',
            'next': 5
        },
        {
            'sequence': [1, 2, 1, 2, 1, '?'],
            'pattern': '1, 2 ÈáçÂ§ç',
            'next': 2
        }
    ]
    
    score = 0
    
    for i, pattern in enumerate(patterns):
        print(f"\nÈóÆÈ¢ò {i+1}:")
        print(f"Êï∞ÂàóÔºö{', '.join(map(str, pattern['sequence']))}")
        print(f"ËßÑÂæãÊòØ‰ªÄ‰πàÔºü")
        
        answer = input("‰Ω†ÁöÑÁ≠îÊ°àÔºö")
        
        if answer == pattern['pattern']:
            print("‚úÖ Á≠îÂØπ‰∫ÜÔºÅ")
            print(f"‰∏ã‰∏Ä‰∏™Êï∞ÊòØÔºö{pattern['next']}")
            score += 1
        else:
            print(f"‚ùå ‰∏çÂØπÂì¶ÔºåËßÑÂæãÊòØÔºö{pattern['pattern']}")
            print(f"‰∏ã‰∏Ä‰∏™Êï∞ÊòØÔºö{pattern['next']}")
    
    print(f"\nüéØ ‰Ω†ÁöÑÂæóÂàÜÔºö{score}/{len(patterns)}")
    
    if score == len(patterns):
        print("üèÜ Â§™ÂéâÂÆ≥‰∫ÜÔºÅÂÖ®ÈÉ®Ê≠£Á°ÆÔºÅ")
    elif score >= len(patterns) * 0.8:
        print("üåü ÂÅöÂæóÂæàÂ•ΩÔºÅÁªßÁª≠‰øùÊåÅÔºÅ")
    else:
        print("üí™ ÁªßÁª≠Âä†Ê≤πÔºåÂ§öÁªÉ‰π†Â∞±‰ºöËøõÊ≠•ÔºÅ")

# ËøêË°åÊâæÊï∞Â≠óËßÑÂæã
find_number_pattern()
```

```python
def find_shape_pattern():
    """ÊâæÂõæÂΩ¢ËßÑÂæã"""
    print("üî∑ ÊâæÂõæÂΩ¢ËßÑÂæã")
    print("=" * 40)
    
    # ÂàõÂª∫ÂõæÂΩ¢ËßÑÂæã
    fig, axes = plt.subplots(3, 2, figsize=(12, 15))
    
    # ËßÑÂæã1ÔºöÂúÜÂΩ¢„ÄÅÊ≠£ÊñπÂΩ¢ÈáçÂ§ç
    ax = axes[0, 0]
    for i in range(5):
        if i % 2 == 0:
            circle = plt.Circle((i, 0), 0.3, facecolor='lightblue', 
                               edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
        else:
            square = plt.Rectangle((i-0.25, -0.25), 0.5, 0.5, facecolor='lightgreen', 
                                  edgecolor='green', linewidth=2)
            ax.add_patch(square)
    ax.text(2, 1, 'ËßÑÂæãÔºöÂúÜÂΩ¢„ÄÅÊ≠£ÊñπÂΩ¢ÈáçÂ§ç', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # ËßÑÂæã2Ôºö‰∏âËßíÂΩ¢„ÄÅÈïøÊñπÂΩ¢ÈáçÂ§ç
    ax = axes[0, 1]
    for i in range(5):
        if i % 2 == 0:
            triangle = plt.Polygon([(i-0.3, -0.25), (i+0.3, -0.25), (i, 0.25)], 
                                  facecolor='lightyellow', edgecolor='orange', linewidth=2)
            ax.add_patch(triangle)
        else:
            rectangle = plt.Rectangle((i-0.4, -0.25), 0.8, 0.5, facecolor='lightpink', 
                                     edgecolor='red', linewidth=2)
            ax.add_patch(rectangle)
    ax.text(2, 1, 'ËßÑÂæãÔºö‰∏âËßíÂΩ¢„ÄÅÈïøÊñπÂΩ¢ÈáçÂ§ç', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # ËßÑÂæã3ÔºöÈ¢úËâ≤ÂèòÂåñ
    ax = axes[1, 0]
    colors = ['red', 'blue', 'green', 'red', 'blue']
    for i, color in enumerate(colors):
        circle = plt.Circle((i, 0), 0.3, facecolor=color, edgecolor=color, linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, 'ËßÑÂæãÔºöÁ∫¢„ÄÅËìù„ÄÅÁªøÈáçÂ§ç', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # ËßÑÂæã4ÔºöÂ§ßÂ∞èÂèòÂåñ
    ax = axes[1, 1]
    for i in range(5):
        size = 0.2 + i * 0.1
        circle = plt.Circle((i, 0), size, facecolor='lightblue', 
                           edgecolor='blue', linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, 'ËßÑÂæãÔºöË∂äÊù•Ë∂äÂ§ß', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # ËßÑÂæã5Ôºö‰ΩçÁΩÆÂèòÂåñ
    ax = axes[2, 0]
    for i in range(5):
        y = 0.3 if i % 2 == 0 else -0.3
        circle = plt.Circle((i, y), 0.3, facecolor='lightgreen', 
                           edgecolor='green', linewidth=2)
        ax.add_patch(circle)
    ax.text(2, 1, 'ËßÑÂæãÔºö‰∏ä‰∏ã‰∫§Êõø', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # ËßÑÂæã6ÔºöÊï∞ÈáèÂèòÂåñ
    ax = axes[2, 1]
    for i in range(5):
        for j in range(i + 1):
            circle = plt.Circle((i, j * 0.3 - 0.6), 0.1, facecolor='lightpink', 
                               edgecolor='red', linewidth=2)
            ax.add_patch(circle)
    ax.text(2, 1, 'ËßÑÂæãÔºöÊï∞ÈáèÈÄíÂ¢û', fontsize=12, ha='center', fontweight='bold')
    ax.text(2, -1, '‰∏ã‰∏Ä‰∏™ÊòØÔºü', fontsize=12, ha='center', fontweight='bold')
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.suptitle('ÊâæÂõæÂΩ¢ËßÑÂæã', fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # ÈóÆÁ≠î
    print("\nü§î ÊÄùËÄÉÈ¢òÔºö")
    print("1. ÂúÜÂΩ¢„ÄÅÊ≠£ÊñπÂΩ¢ÈáçÂ§çÔºå‰∏ã‰∏Ä‰∏™ÊòØ‰ªÄ‰πàÔºü")
    print("   Á≠îÊ°àÔºöÂúÜÂΩ¢")
    print("2. ‰∏âËßíÂΩ¢„ÄÅÈïøÊñπÂΩ¢ÈáçÂ§çÔºå‰∏ã‰∏Ä‰∏™ÊòØ‰ªÄ‰πàÔºü")
    print("   Á≠îÊ°àÔºö‰∏âËßíÂΩ¢")
    print("3. Á∫¢„ÄÅËìù„ÄÅÁªøÈáçÂ§çÔºå‰∏ã‰∏Ä‰∏™ÊòØ‰ªÄ‰πàÈ¢úËâ≤Ôºü")
    print("   Á≠îÊ°àÔºöÁªøËâ≤")
    print("4. Ë∂äÊù•Ë∂äÂ§ßÔºå‰∏ã‰∏Ä‰∏™ÊòØ‰ªÄ‰πàÔºü")
    print("   Á≠îÊ°àÔºöÊõ¥Â§ßÁöÑÂúÜ")
    print("5. ‰∏ä‰∏ã‰∫§ÊõøÔºå‰∏ã‰∏Ä‰∏™Âú®Âì™ÈáåÔºü")
    print("   Á≠îÊ°àÔºö‰∏äÈù¢")
    print("6. Êï∞ÈáèÈÄíÂ¢ûÔºå‰∏ã‰∏Ä‰∏™ÊúâÂá†‰∏™Ôºü")
    print("   Á≠îÊ°àÔºö5‰∏™")

# ËøêË°åÊâæÂõæÂΩ¢ËßÑÂæã
find_shape_pattern()
```

## 5. ÊâæËßÑÂæãÊ∏∏Êàè

Êù•Áé©ÊâæËßÑÂæãÁöÑÊ∏∏ÊàèÂêßÔºÅ

```python
import random

def pattern_game():
    """ÊâæËßÑÂæãÊ∏∏Êàè"""
    print("üéÆ ÊâæËßÑÂæãÊ∏∏Êàè")
    print("=" * 40)
    print("ËßÑÂàôÔºöÁúãËßÑÂæãÔºåËØ¥Âá∫‰∏ã‰∏Ä‰∏™ÊòØ‰ªÄ‰πà")
    
    score = 0
    total_questions = 5
    
    # ÂàõÂª∫ÂêÑÁßçËßÑÂæã
    patterns = [
        {
            'type': 'number',
            'sequence': [1, 2, 3, 4, 5],
            'pattern': 'ÊØèÊ¨°Âä†1',
            'next': 6
        },
        {
            'type': 'number',
            'sequence': [2, 4, 6, 8, 10],
            'pattern': 'ÊØèÊ¨°Âä†2',
            'next': 12
        },
        {
            'type': 'number',
            'sequence': [10, 9, 8, 7, 6],
            'pattern': 'ÊØèÊ¨°Âáè1',
            'next': 5
        },
        {
            'type': 'number',
            'sequence': [1, 2, 1, 2, 1],
            'pattern': '1, 2ÈáçÂ§ç',
            'next': 2
        },
        {
            'type': 'number',
            'sequence': [1, 1, 2, 2, 3],
            'pattern': 'ÊØè‰∏™Êï∞ÈáçÂ§ç2Ê¨°',
            'next': 3
        }
    ]
    
    # ÈöèÊú∫ÈÄâÊã©5‰∏™ÈóÆÈ¢ò
    selected_patterns = random.sample(patterns, min(total_questions, len(patterns)))
    
    for i, pattern in enumerate(selected_patterns):
        print(f"\nÈóÆÈ¢ò {i+1}/{total_questions}")
        
        # ÊòæÁ§∫ËßÑÂæã
        print(f"Êï∞ÂàóÔºö{', '.join(map(str, pattern['sequence']))}")
        
        # ËØ¢ÈóÆËßÑÂæã
        pattern_answer = input("ËßÑÂæãÊòØ‰ªÄ‰πàÔºü")
        
        if pattern_answer == pattern['pattern']:
            print("‚úÖ ËßÑÂæãÁ≠îÂØπ‰∫ÜÔºÅ")
            
            # ËØ¢ÈóÆ‰∏ã‰∏Ä‰∏™
            next_answer = input(f"‰∏ã‰∏Ä‰∏™Êï∞ÊòØ‰ªÄ‰πàÔºü")
            
            try:
                if int(next_answer) == pattern['next']:
                    print("‚úÖ ÂÖ®ÈÉ®Ê≠£Á°ÆÔºÅ")
                    score += 1
                else:
                    print(f"‚ùå ‰∏çÂØπÂì¶Ôºå‰∏ã‰∏Ä‰∏™Êï∞ÊòØ {pattern['next']}")
            except:
                print(f"‚ùå ‰∏çÂØπÂì¶Ôºå‰∏ã‰∏Ä‰∏™Êï∞ÊòØ {pattern['next']}")
        else:
            print(f"‚ùå ‰∏çÂØπÂì¶ÔºåËßÑÂæãÊòØÔºö{pattern['pattern']}")
            print(f"‰∏ã‰∏Ä‰∏™Êï∞ÊòØÔºö{pattern['next']}")
    
    print(f"\nüéØ Ê∏∏ÊàèÁªìÊùüÔºÅ")
    print(f"‰Ω†ÁöÑÂæóÂàÜÔºö{score}/{total_questions}")
    
    if score == total_questions:
        print("üèÜ Â§™ÂéâÂÆ≥‰∫ÜÔºÅÂÖ®ÈÉ®Ê≠£Á°ÆÔºÅ")
    elif score >= total_questions * 0.8:
        print("üåü ÂÅöÂæóÂæàÂ•ΩÔºÅÁªßÁª≠‰øùÊåÅÔºÅ")
    else:
        print("üí™ ÁªßÁª≠Âä†Ê≤πÔºåÂ§öÁªÉ‰π†Â∞±‰ºöËøõÊ≠•ÔºÅ")

# ËøêË°åÊ∏∏Êàè
pattern_game()
```
