## 1. è®¤è¯†åŸºæœ¬å›¾å½¢

è®©æˆ‘ä»¬è®¤è¯†ä¸€äº›åŸºæœ¬çš„å›¾å½¢ï¼

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_basic_shapes():
    """ç»˜åˆ¶åŸºæœ¬å›¾å½¢"""
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    shapes = [
        {'name': 'åœ†å½¢', 'color': 'lightblue', 'edge': 'blue'},
        {'name': 'æ­£æ–¹å½¢', 'color': 'lightgreen', 'edge': 'green'},
        {'name': 'ä¸‰è§’å½¢', 'color': 'lightyellow', 'edge': 'orange'},
        {'name': 'é•¿æ–¹å½¢', 'color': 'lightpink', 'edge': 'red'},
        {'name': 'æ¤­åœ†å½¢', 'color': 'lightcyan', 'edge': 'cyan'},
        {'name': 'è±å½¢', 'color': 'lavender', 'edge': 'purple'}
    ]
    
    for i, (ax, shape) in enumerate(zip(axes, shapes)):
        ax.set_xlim(-1.5, 1.5)
        ax.set_ylim(-1.5, 1.5)
        ax.set_aspect('equal')
        ax.axis('off')
        ax.set_title(shape['name'], fontsize=16, fontweight='bold', pad=20)
        
        if i == 0:  # åœ†å½¢
            circle = plt.Circle((0, 0), 1, facecolor=shape['color'], 
                               edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(circle)
        elif i == 1:  # æ­£æ–¹å½¢
            square = plt.Rectangle((-0.8, -0.8), 1.6, 1.6, 
                                  facecolor=shape['color'], edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(square)
        elif i == 2:  # ä¸‰è§’å½¢
            triangle = plt.Polygon([(-0.8, -0.8), (0.8, -0.8), (0, 0.8)], 
                                   facecolor=shape['color'], edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(triangle)
        elif i == 3:  # é•¿æ–¹å½¢
            rectangle = plt.Rectangle((-1.0, -0.6), 2.0, 1.2, 
                                     facecolor=shape['color'], edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(rectangle)
        elif i == 4:  # æ¤­åœ†å½¢
            ellipse = plt.Ellipse((0, 0), 2.0, 1.2, 
                                  facecolor=shape['color'], edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(ellipse)
        elif i == 5:  # è±å½¢
            diamond = plt.Polygon([(0, 1), (0.8, 0), (0, -1), (-0.8, 0)], 
                                   facecolor=shape['color'], edgecolor=shape['edge'], linewidth=3)
            ax.add_patch(diamond)
    
    plt.suptitle('è®¤è¯†åŸºæœ¬å›¾å½¢', fontsize=20, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()

# æ˜¾ç¤ºåŸºæœ¬å›¾å½¢
print("ğŸ”· è®¤è¯†åŸºæœ¬å›¾å½¢ï¼š")
draw_basic_shapes()
```

## 2. å›¾å½¢æ¨¡å¼è¯†åˆ«

æ‰¾å‡ºå›¾å½¢çš„è§„å¾‹ï¼ŒçŒœå‡ºä¸‹ä¸€ä¸ªæ˜¯ä»€ä¹ˆï¼

```python
def draw_pattern_sequence(pattern_type):
    """ç»˜åˆ¶å›¾å½¢æ¨¡å¼åºåˆ—"""
    fig, ax = plt.subplots(figsize=(14, 4))
    
    patterns = {
        'circle': {'shape': 'circle', 'colors': ['lightblue', 'lightgreen', 'lightyellow']},
        'square': {'shape': 'square', 'colors': ['lightcoral', 'lightpink', 'lightcyan']},
        'triangle': {'shape': 'triangle', 'colors': ['lightyellow', 'lightblue', 'lightgreen']},
        'mixed': {'shape': 'mixed', 'colors': ['lightblue', 'lightgreen', 'lightyellow']}
    }
    
    pattern = patterns[pattern_type]
    colors = pattern['colors']
    
    # ç»˜åˆ¶6ä¸ªå›¾å½¢
    for i in range(6):
        x = i * 2.0 + 1.0
        y = 0.5
        color = colors[i % len(colors)]
        
        if pattern_type == 'circle':
            shape = plt.Circle((x, y), 0.6, facecolor=color, 
                             edgecolor='black', linewidth=2)
        elif pattern_type == 'square':
            shape = plt.Rectangle((x-0.5, y-0.5), 1.0, 1.0, 
                                 facecolor=color, edgecolor='black', linewidth=2)
        elif pattern_type == 'triangle':
            shape = plt.Polygon([(x, y+0.6), (x-0.5, y-0.4), (x+0.5, y-0.4)], 
                               facecolor=color, edgecolor='black', linewidth=2)
        elif pattern_type == 'mixed':
            if i % 3 == 0:
                shape = plt.Circle((x, y), 0.6, facecolor=color, 
                                 edgecolor='black', linewidth=2)
            elif i % 3 == 1:
                shape = plt.Rectangle((x-0.5, y-0.5), 1.0, 1.0, 
                                     facecolor=color, edgecolor='black', linewidth=2)
            else:
                shape = plt.Polygon([(x, y+0.6), (x-0.5, y-0.4), (x+0.5, y-0.4)], 
                                   facecolor=color, edgecolor='black', linewidth=2)
        
        ax.add_patch(shape)
        ax.text(x, y-0.8, str(i+1), fontsize=14, ha='center', fontweight='bold')
    
    # é—®å·
    ax.text(13.0, 0.5, '?', fontsize=40, ha='center', va='center', 
            fontweight='bold', color='red')
    
    ax.set_xlim(0, 14)
    ax.set_ylim(0, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    
    title = f'å›¾å½¢æ¨¡å¼ï¼š{pattern_type}'
    ax.set_title(title, fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# å±•ç¤ºä¸åŒçš„å›¾å½¢æ¨¡å¼
print("ğŸ” å›¾å½¢æ¨¡å¼è¯†åˆ«ï¼š")
print("\næ¨¡å¼1ï¼šé¢œè‰²å¾ªç¯")
draw_pattern_sequence('circle')
print("è§„å¾‹ï¼šé¢œè‰²æŒ‰è“ã€ç»¿ã€é»„å¾ªç¯")

print("\næ¨¡å¼2ï¼šå½¢çŠ¶ç›¸åŒï¼Œé¢œè‰²å˜åŒ–")
draw_pattern_sequence('square')
print("è§„å¾‹ï¼šéƒ½æ˜¯æ­£æ–¹å½¢ï¼Œé¢œè‰²æŒ‰çº¢ã€ç²‰ã€é’å¾ªç¯")

print("\næ¨¡å¼3ï¼šå½¢çŠ¶å’Œé¢œè‰²éƒ½å˜åŒ–")
draw_pattern_sequence('mixed')
print("è§„å¾‹ï¼šåœ†å½¢ã€æ­£æ–¹å½¢ã€ä¸‰è§’å½¢å¾ªç¯ï¼Œé¢œè‰²ä¹Ÿå¾ªç¯")
```

## 3. æ‰¾è§„å¾‹å¡«ç©º

æ‰¾å‡ºå›¾å½¢åºåˆ—çš„è§„å¾‹ï¼Œå¡«ä¸Šæ­£ç¡®çš„å›¾å½¢ï¼

```python
def pattern_game():
    """å›¾å½¢è§„å¾‹æ¸¸æˆ"""
    print("ğŸ® å›¾å½¢è§„å¾‹æ¸¸æˆ")
    print("=" * 40)
    
    patterns = [
        {
            'sequence': ['åœ†å½¢', 'æ­£æ–¹å½¢', 'åœ†å½¢', 'æ­£æ–¹å½¢', '?'],
            'rule': 'åœ†å½¢å’Œæ­£æ–¹å½¢äº¤æ›¿',
            'answer': 'åœ†å½¢'
        },
        {
            'sequence': ['çº¢è‰²', 'è“è‰²', 'ç»¿è‰²', 'çº¢è‰²', 'è“è‰²', '?'],
            'rule': 'çº¢ã€è“ã€ç»¿ä¸‰ç§é¢œè‰²å¾ªç¯',
            'answer': 'ç»¿è‰²'
        },
        {
            'sequence': ['1ä¸ªåœ†', '2ä¸ªåœ†', '3ä¸ªåœ†', '4ä¸ªåœ†', '?'],
            'rule': 'æ¯æ¬¡å¢åŠ 1ä¸ªåœ†',
            'answer': '5ä¸ªåœ†'
        },
        {
            'sequence': ['å¤§åœ†', 'å°åœ†', 'å¤§åœ†', 'å°åœ†', '?'],
            'rule': 'å¤§åœ†å’Œå°åœ†äº¤æ›¿',
            'answer': 'å¤§åœ†'
        },
        {
            'sequence': ['ä¸‰è§’å½¢', 'æ­£æ–¹å½¢', 'äº”è¾¹å½¢', 'ä¸‰è§’å½¢', 'æ­£æ–¹å½¢', '?'],
            'rule': 'ä¸‰è§’å½¢ã€æ­£æ–¹å½¢ã€äº”è¾¹å½¢å¾ªç¯',
            'answer': 'äº”è¾¹å½¢'
        }
    ]
    
    score = 0
    
    for i, pattern in enumerate(patterns):
        print(f"\né—®é¢˜ {i+1}:")
        seq_str = ', '.join(pattern['sequence'])
        print(f"åºåˆ—ï¼š{seq_str}")
        print(f"æç¤ºï¼š{pattern['rule']}")
        
        answer = input("é—®å·å¤„åº”è¯¥æ˜¯ä»€ä¹ˆï¼Ÿ")
        
        if answer == pattern['answer']:
            print("âœ… ç­”å¯¹äº†ï¼")
            score += 1
        else:
            print(f"âŒ ä¸å¯¹å“¦ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {pattern['answer']}")
    
    print(f"\nğŸ¯ æ¸¸æˆç»“æŸï¼")
    print(f"ä½ çš„å¾—åˆ†ï¼š{score}/{len(patterns)}")
    
    if score == len(patterns):
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif score >= len(patterns) * 0.8:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šè§‚å¯Ÿå°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œæ¸¸æˆ
pattern_game()
```

## 4. å›¾å½¢è®¡æ•°

æ•°ä¸€æ•°å›¾ä¸­æœ‰å¤šå°‘ä¸ªå›¾å½¢ï¼

```python
def count_shapes():
    """å›¾å½¢è®¡æ•°"""
    import random
    
    print("ğŸ”¢ å›¾å½¢è®¡æ•°")
    print("=" * 40)
    
    # ç”Ÿæˆéšæœºå›¾å½¢
    num_circles = random.randint(3, 6)
    num_squares = random.randint(3, 6)
    num_triangles = random.randint(3, 6)
    
    # åˆ›å»ºå›¾å½¢åˆ—è¡¨
    shapes = []
    for i in range(num_circles):
        shapes.append({'type': 'circle', 'color': 'lightblue', 'edge': 'blue'})
    for i in range(num_squares):
        shapes.append({'type': 'square', 'color': 'lightgreen', 'edge': 'green'})
    for i in range(num_triangles):
        shapes.append({'type': 'triangle', 'color': 'lightyellow', 'edge': 'orange'})
    
    # æ‰“ä¹±é¡ºåº
    random.shuffle(shapes)
    
    # ç»˜åˆ¶å›¾å½¢
    fig, ax = plt.subplots(figsize=(14, 8))
    
    cols = 5
    rows = (len(shapes) + cols - 1) // cols
    
    for i, shape in enumerate(shapes):
        row = i // cols
        col = i % cols
        x = col * 2.5 + 1.25
        y = (rows - row - 1) * 2.0 + 1.0
        
        if shape['type'] == 'circle':
            patch = plt.Circle((x, y), 0.6, facecolor=shape['color'], 
                             edgecolor=shape['edge'], linewidth=2)
        elif shape['type'] == 'square':
            patch = plt.Rectangle((x-0.5, y-0.5), 1.0, 1.0, 
                                 facecolor=shape['color'], edgecolor=shape['edge'], linewidth=2)
        elif shape['type'] == 'triangle':
            patch = plt.Polygon([(x, y+0.6), (x-0.5, y-0.4), (x+0.5, y-0.4)], 
                               facecolor=shape['color'], edgecolor=shape['edge'], linewidth=2)
        
        ax.add_patch(patch)
        ax.text(x, y-0.8, str(i+1), fontsize=12, ha='center', fontweight='bold')
    
    ax.set_xlim(0, cols * 2.5)
    ax.set_ylim(0, rows * 2.0 + 1.0)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('æ•°ä¸€æ•°æœ‰å¤šå°‘ä¸ªå›¾å½¢', fontsize=18, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()
    
    # è¯¢é—®ç­”æ¡ˆ
    print(f"\nè¯·æ•°ä¸€æ•°ï¼š")
    circle_answer = input("æœ‰å¤šå°‘ä¸ªåœ†å½¢ï¼Ÿ")
    square_answer = input("æœ‰å¤šå°‘ä¸ªæ­£æ–¹å½¢ï¼Ÿ")
    triangle_answer = input("æœ‰å¤šå°‘ä¸ªä¸‰è§’å½¢ï¼Ÿ")
    total_answer = input("ä¸€å…±æœ‰å¤šå°‘ä¸ªå›¾å½¢ï¼Ÿ")
    
    # æ£€æŸ¥ç­”æ¡ˆ
    correct = 0
    total = 4
    
    try:
        if int(circle_answer) == num_circles:
            print(f"âœ… åœ†å½¢æ•°é‡æ­£ç¡®ï¼")
            correct += 1
        else:
            print(f"âŒ åœ†å½¢æ•°é‡ä¸å¯¹ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {num_circles}")
    except:
        print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œåœ†å½¢æ•°é‡æ˜¯ {num_circles}")
    
    try:
        if int(square_answer) == num_squares:
            print(f"âœ… æ­£æ–¹å½¢æ•°é‡æ­£ç¡®ï¼")
            correct += 1
        else:
            print(f"âŒ æ­£æ–¹å½¢æ•°é‡ä¸å¯¹ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {num_squares}")
    except:
        print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œæ­£æ–¹å½¢æ•°é‡æ˜¯ {num_squares}")
    
    try:
        if int(triangle_answer) == num_triangles:
            print(f"âœ… ä¸‰è§’å½¢æ•°é‡æ­£ç¡®ï¼")
            correct += 1
        else:
            print(f"âŒ ä¸‰è§’å½¢æ•°é‡ä¸å¯¹ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {num_triangles}")
    except:
        print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œä¸‰è§’å½¢æ•°é‡æ˜¯ {num_triangles}")
    
    try:
        if int(total_answer) == len(shapes):
            print(f"âœ… æ€»æ•°é‡æ­£ç¡®ï¼")
            correct += 1
        else:
            print(f"âŒ æ€»æ•°é‡ä¸å¯¹ï¼Œæ­£ç¡®ç­”æ¡ˆæ˜¯ {len(shapes)}")
    except:
        print(f"âŒ è¯·è¾“å…¥æ•°å­—ï¼Œæ€»æ•°é‡æ˜¯ {len(shapes)}")
    
    print(f"\nğŸ¯ ä½ çš„å¾—åˆ†ï¼š{correct}/{total}")
    
    if correct == total:
        print("ğŸ† å¤ªå‰å®³äº†ï¼å…¨éƒ¨æ­£ç¡®ï¼")
    elif correct >= total * 0.75:
        print("ğŸŒŸ åšå¾—å¾ˆå¥½ï¼ç»§ç»­ä¿æŒï¼")
    else:
        print("ğŸ’ª ç»§ç»­åŠ æ²¹ï¼Œå¤šè§‚å¯Ÿå°±ä¼šè¿›æ­¥ï¼")

# è¿è¡Œè®¡æ•°æ¸¸æˆ
count_shapes()
```

## 5. å›¾å½¢æ‹¼å›¾

```python
def create_puzzle():
    """åˆ›å»ºå›¾å½¢æ‹¼å›¾"""
    fig, axes = plt.subplots(2, 2, figsize=(14, 14))
    axes = axes.flatten()
    
    # æ‹¼å›¾1ï¼šå°æˆ¿å­
    ax = axes[0]
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('å°æˆ¿å­', fontsize=16, fontweight='bold', pad=20)
    
    # å±‹é¡¶ï¼ˆä¸‰è§’å½¢ï¼‰
    roof = plt.Polygon([(2, 8), (5, 10), (8, 8)], 
                      facecolor='lightcoral', edgecolor='red', linewidth=2)
    ax.add_patch(roof)
    
    # æˆ¿å­ä¸»ä½“ï¼ˆæ­£æ–¹å½¢ï¼‰
    house = plt.Rectangle((2.5, 4), 5, 4, 
                         facecolor='lightblue', edgecolor='blue', linewidth=2)
    ax.add_patch(house)
    
    # é—¨ï¼ˆé•¿æ–¹å½¢ï¼‰
    door = plt.Rectangle((4, 4), 2, 2.5, 
                        facecolor='lightyellow', edgecolor='orange', linewidth=2)
    ax.add_patch(door)
    
    # çª—æˆ·ï¼ˆåœ†å½¢ï¼‰
    window = plt.Circle((3.5, 6), 0.5, facecolor='lightgreen', 
                       edgecolor='green', linewidth=2)
    ax.add_patch(window)
    
    # æ‹¼å›¾2ï¼šå°æ ‘
    ax = axes[1]
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('å°æ ‘', fontsize=16, fontweight='bold', pad=20)
    
    # æ ‘å¹²ï¼ˆé•¿æ–¹å½¢ï¼‰
    trunk = plt.Rectangle((4.5, 2), 1, 3, 
                         facecolor='lightbrown', edgecolor='brown', linewidth=2)
    ax.add_patch(trunk)
    
    # æ ‘å¶ï¼ˆä¸‰è§’å½¢ï¼‰
    leaves1 = plt.Polygon([(2, 5), (5, 8), (8, 5)], 
                         facecolor='lightgreen', edgecolor='green', linewidth=2)
    ax.add_patch(leaves1)
    
    leaves2 = plt.Polygon([(2.5, 6.5), (5, 9), (7.5, 6.5)], 
                         facecolor='lightgreen', edgecolor='green', linewidth=2)
    ax.add_patch(leaves2)
    
    # æ‹¼å›¾3ï¼šå°èˆ¹
    ax = axes[2]
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('å°èˆ¹', fontsize=16, fontweight='bold', pad=20)
    
    # èˆ¹èº«ï¼ˆåŠåœ†å½¢ï¼‰
    boat = plt.Polygon([(2, 4), (8, 4), (7, 6), (3, 6)], 
                      facecolor='lightblue', edgecolor='blue', linewidth=2)
    ax.add_patch(boat)
    
    # å¸†ï¼ˆä¸‰è§’å½¢ï¼‰
    sail = plt.Polygon([(5, 4), (5, 9), (7, 6)], 
                      facecolor='lightyellow', edgecolor='orange', linewidth=2)
    ax.add_patch(sail)
    
    # æ‹¼å›¾4ï¼šå¤ªé˜³
    ax = axes[3]
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('å¤ªé˜³', fontsize=16, fontweight='bold', pad=20)
    
    # å¤ªé˜³ä¸»ä½“ï¼ˆåœ†å½¢ï¼‰
    sun = plt.Circle((5, 5), 2, facecolor='lightyellow', 
                    edgecolor='orange', linewidth=3)
    ax.add_patch(sun)
    
    # å¤ªé˜³å…‰èŠ’ï¼ˆçº¿æ¡ï¼‰
    for i in range(8):
        angle = i * np.pi / 4
        x1 = 5 + 2.2 * np.cos(angle)
        y1 = 5 + 2.2 * np.sin(angle)
        x2 = 5 + 3.0 * np.cos(angle)
        y2 = 5 + 3.0 * np.sin(angle)
        ax.plot([x1, x2], [y1, y2], 'orange', linewidth=3)
    
    plt.suptitle('å›¾å½¢æ‹¼å›¾', fontsize=20, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()
    
    print("ğŸ§© å›¾å½¢æ‹¼å›¾")
    print("=" * 40)
    print("\nè¿™äº›å›¾æ¡ˆéƒ½æ˜¯ç”±åŸºæœ¬å›¾å½¢æ‹¼æˆçš„ï¼š")
    print("- å°æˆ¿å­ï¼šä¸‰è§’å½¢ + æ­£æ–¹å½¢ + é•¿æ–¹å½¢ + åœ†å½¢")
    print("- å°æ ‘ï¼šé•¿æ–¹å½¢ + ä¸‰è§’å½¢")
    print("- å°èˆ¹ï¼šå››è¾¹å½¢ + ä¸‰è§’å½¢")
    print("- å¤ªé˜³ï¼šåœ†å½¢ + çº¿æ¡")
    print("\nä½ ä¹Ÿå¯ä»¥ç”¨åŸºæœ¬å›¾å½¢æ‹¼å‡ºæœ‰è¶£çš„å›¾æ¡ˆï¼")

# æ˜¾ç¤ºæ‹¼å›¾
create_puzzle()
```

## 6. å›¾å½¢åˆ†ç±»

å­¦ä¹ å¦‚ä½•æŒ‰ä¸åŒçš„æ ‡å‡†ç»™å›¾å½¢åˆ†ç±»ï¼

```python
def classify_shapes():
    """å›¾å½¢åˆ†ç±»"""
    print("ğŸ“Š å›¾å½¢åˆ†ç±»")
    print("=" * 40)
    
    # åˆ›å»ºä¸åŒå½¢çŠ¶å’Œé¢œè‰²çš„å›¾å½¢
    shapes = [
        {'type': 'circle', 'size': 'big', 'color': 'red'},
        {'type': 'circle', 'size': 'small', 'color': 'blue'},
        {'type': 'square', 'size': 'big', 'color': 'green'},
        {'type': 'square', 'size': 'small', 'color': 'yellow'},
        {'type': 'triangle', 'size': 'big', 'color': 'red'},
        {'type': 'triangle', 'size': 'small', 'color': 'blue'}
    ]
    
    # æŒ‰å½¢çŠ¶åˆ†ç±»
    print("\n1. æŒ‰å½¢çŠ¶åˆ†ç±»ï¼š")
    circles = [s for s in shapes if s['type'] == 'circle']
    squares = [s for s in shapes if s['type'] == 'square']
    triangles = [s for s in shapes if s['type'] == 'triangle']
    
    print(f"   åœ†å½¢ï¼š{len(circles)} ä¸ª")
    print(f"   æ­£æ–¹å½¢ï¼š{len(squares)} ä¸ª")
    print(f"   ä¸‰è§’å½¢ï¼š{len(triangles)} ä¸ª")
    
    # æŒ‰å¤§å°åˆ†ç±»
    print("\n2. æŒ‰å¤§å°åˆ†ç±»ï¼š")
    big_shapes = [s for s in shapes if s['size'] == 'big']
    small_shapes = [s for s in shapes if s['size'] == 'small']
    
    print(f"   å¤§å›¾å½¢ï¼š{len(big_shapes)} ä¸ª")
    print(f"   å°å›¾å½¢ï¼š{len(small_shapes)} ä¸ª")
    
    # æŒ‰é¢œè‰²åˆ†ç±»
    print("\n3. æŒ‰é¢œè‰²åˆ†ç±»ï¼š")
    red_shapes = [s for s in shapes if s['color'] == 'red']
    blue_shapes = [s for s in shapes if s['color'] == 'blue']
    green_shapes = [s for s in shapes if s['color'] == 'green']
    yellow_shapes = [s for s in shapes if s['color'] == 'yellow']
    
    print(f"   çº¢è‰²ï¼š{len(red_shapes)} ä¸ª")
    print(f"   è“è‰²ï¼š{len(blue_shapes)} ä¸ª")
    print(f"   ç»¿è‰²ï¼š{len(green_shapes)} ä¸ª")
    print(f"   é»„è‰²ï¼š{len(yellow_shapes)} ä¸ª")
    
    # å¯è§†åŒ–
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # æŒ‰å½¢çŠ¶åˆ†ç±»çš„å¯è§†åŒ–
    ax = axes[0]
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('æŒ‰å½¢çŠ¶åˆ†ç±»', fontsize=14, fontweight='bold', pad=20)
    
    for i, shape in enumerate(circles):
        size = 0.5 if shape['size'] == 'small' else 0.7
        color_map = {'red': 'lightcoral', 'blue': 'lightblue', 'green': 'lightgreen', 'yellow': 'lightyellow'}
        circle = plt.Circle((0.5 + i*0.8, 1.5), size, 
                           facecolor=color_map[shape['color']], edgecolor='black', linewidth=2)
        ax.add_patch(circle)
    
    for i, shape in enumerate(squares):
        size = 0.4 if shape['size'] == 'small' else 0.6
        color_map = {'red': 'lightcoral', 'blue': 'lightblue', 'green': 'lightgreen', 'yellow': 'lightyellow'}
        square = plt.Rectangle((0.5 + i*0.8 - size, 0.5 - size), size*2, size*2, 
                              facecolor=color_map[shape['color']], edgecolor='black', linewidth=2)
        ax.add_patch(square)
    
    # æŒ‰å¤§å°åˆ†ç±»çš„å¯è§†åŒ–
    ax = axes[1]
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('æŒ‰å¤§å°åˆ†ç±»', fontsize=14, fontweight='bold', pad=20)
    
    for i, shape in enumerate(big_shapes):
        if shape['type'] == 'circle':
            patch = plt.Circle((0.5 + i*0.8, 2.0), 0.7, 
                             facecolor='lightgray', edgecolor='black', linewidth=2)
        elif shape['type'] == 'square':
            patch = plt.Rectangle((0.5 + i*0.8 - 0.6, 2.0 - 0.6), 1.2, 1.2, 
                                 facecolor='lightgray', edgecolor='black', linewidth=2)
        elif shape['type'] == 'triangle':
            patch = plt.Polygon([(0.5 + i*0.8, 2.7), (0.5 + i*0.8 - 0.6, 2.0 - 0.6), 
                               (0.5 + i*0.8 + 0.6, 2.0 - 0.6)], 
                              facecolor='lightgray', edgecolor='black', linewidth=2)
        ax.add_patch(patch)
    
    for i, shape in enumerate(small_shapes):
        if shape['type'] == 'circle':
            patch = plt.Circle((0.5 + i*0.8, 0.7), 0.5, 
                             facecolor='lightgray', edgecolor='black', linewidth=2)
        elif shape['type'] == 'square':
            patch = plt.Rectangle((0.5 + i*0.8 - 0.4, 0.7 - 0.4), 0.8, 0.8, 
                                 facecolor='lightgray', edgecolor='black', linewidth=2)
        elif shape['type'] == 'triangle':
            patch = plt.Polygon([(0.5 + i*0.8, 1.2), (0.5 + i*0.8 - 0.4, 0.7 - 0.4), 
                               (0.5 + i*0.8 + 0.4, 0.7 - 0.4)], 
                              facecolor='lightgray', edgecolor='black', linewidth=2)
        ax.add_patch(patch)
    
    # æŒ‰é¢œè‰²åˆ†ç±»çš„å¯è§†åŒ–
    ax = axes[2]
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title('æŒ‰é¢œè‰²åˆ†ç±»', fontsize=14, fontweight='bold', pad=20)
    
    color_positions = {'red': 2.5, 'blue': 1.8, 'green': 1.1, 'yellow': 0.4}
    
    for shape in shapes:
        y = color_positions[shape['color']]
        x = 0.5 + [s for s in shapes if s['color'] == shape['color']].index(shape) * 0.8
        
        if shape['type'] == 'circle':
            size = 0.5 if shape['size'] == 'small' else 0.7
            color_map = {'red': 'lightcoral', 'blue': 'lightblue', 'green': 'lightgreen', 'yellow': 'lightyellow'}
            patch = plt.Circle((x, y), size, 
                             facecolor=color_map[shape['color']], edgecolor='black', linewidth=2)
        elif shape['type'] == 'square':
            size = 0.4 if shape['size'] == 'small' else 0.6
            color_map = {'red': 'lightcoral', 'blue': 'lightblue', 'green': 'lightgreen', 'yellow': 'lightyellow'}
            patch = plt.Rectangle((x - size, y - size), size*2, size*2, 
                                 facecolor=color_map[shape['color']], edgecolor='black', linewidth=2)
        elif shape['type'] == 'triangle':
            size = 0.5 if shape['size'] == 'small' else 0.7
            color_map = {'red': 'lightcoral', 'blue': 'lightblue', 'green': 'lightgreen', 'yellow': 'lightyellow'}
            patch = plt.Polygon([(x, y + size), (x - size, y - size), (x + size, y - size)], 
                              facecolor=color_map[shape['color']], edgecolor='black', linewidth=2)
        ax.add_patch(patch)
    
    plt.tight_layout()
    plt.show()

# è¿è¡Œåˆ†ç±»
classify_shapes()
```
